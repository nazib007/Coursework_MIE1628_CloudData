## Parsing: Seperate Brand and Item

````python
def parse_col(c):
    # c = "QTY_Bx_y" or "PROMO_Bx_Y"
    # we will segregate the x and y using this function
    parts = c.split('_')
    brand = parts[1]  # e.g., "B1"
    item = parts[2]   # e.g., "1"
    return brand, item
````

## Sales in a Specific Time-Frame

````python
import numpy as np
from typing import Tuple, Union

def sum_column_between_dates(
    df: pd.DataFrame,
    column_name: str,
    start_date: Union[str, pd.Timestamp],
    end_date: Union[str, pd.Timestamp],
) -> Tuple[float, pd.Timestamp, pd.Timestamp]:
    """
    Sum `column_name` between the rows whose `date_col` are closest to `start_date` and `end_date`.

    Parameters
    ----------
    df : DataFrame
        Data containing a datetime column.
    column_name : str
        Column to sum (e.g., "QTY_B1_1").
    start_date : str | pd.Timestamp
        Desired start date (will snap to the closest available date in df[date_col]).
    end_date : str | pd.Timestamp
        Desired end date (will snap to the closest available date in df[date_col]).
    date_col : str
        Name of the datetime column (default "DATE").

    Returns
    -------
    total : float
        Sum of the column over the snapped date range (inclusive).
    snapped_start : pd.Timestamp
        The actual start date used (closest match found in the data).
    snapped_end : pd.Timestamp
        The actual end date used (closest match found in the data).
    """

    dates = df.index  # Index is in date-time format
    
    #---------Date Time conversion if needed---------------------------------

    # Convert inputs to Timestamp
    start_ts = pd.to_datetime(start_date)
    end_ts = pd.to_datetime(end_date)
    if pd.isna(start_ts) or pd.isna(end_ts):  # check if the date-time conversion failed
        raise ValueError("start_date and end_date must be valid datetimes.")

    #----------Snapping to closest Start date and end date--------------------

    # Find closest indices by minimizing absolute time difference
    idx_start = np.argmin(np.abs(dates - start_ts))
    idx_end = np.argmin(np.abs(dates - end_ts))

    # Order correctly
    if idx_start > idx_end:
        idx_start, idx_end = idx_end, idx_start

    snapped_start = dates[idx_start]
    snapped_end = dates[idx_end]

    total = df.loc[snapped_start:snapped_end, column_name].sum()
    return float(total), snapped_start, snapped_end
````

````python
def sum_brand_between_dates(
        df,
        brand_name: str ,
        start_date: Union[str, pd.Timestamp],
        end_date: Union[str, pd.Timestamp],
        brand_dict: dict = brand_dict,
        ) -> Tuple[float, pd.Timestamp, pd.Timestamp, float]:
    """Finds the sum of sales of the brand within the given dates
    brand_name = 'B1', 'B2' example"""

    #------Error Check:------------------
    # Ensure brand_name exists in the dictionary
    if brand_name not in brand_dict:
        print(f"Brand {brand_name} not found in brand_dict")
        return 0.0, pd.NaT, pd.NaT, []

    # Compute sales of each item in the given date range
    item_sales = []
    for item in brand_dict[brand_name]:
        sales_qty, start_timestamp, end_timestamp = sum_column_between_dates(df, f'QTY_{brand_name}_{item}', start_date, end_date)
        item_sales.append(sales_qty)
        brand_sales = sum(item_sales)

    return brand_sales, start_timestamp, end_timestamp, item_sales
````

## Bar Plot- Stacked by Year- Items in Each Brand

For each Brand, we will plot all the items of that brand in a tiered bar-plot

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from typing import Optional


def plot_brand_tiered_bars(
    df: pd.DataFrame,
    brand_name: str,
    top_n: Optional[int] = 20,
    rotate_xticks: int = 90,
):
    """
    Tiered (stacked-by-year) bar plot for a brand.
    
    X-axis: item labels (e.g., 1, 2, 3 ... from columns QTY_B{brand}_{item})
    Y-axis: sales quantity (sum over each year)
    Tiers: each stacked layer is a different year
    
    Parameters
    ----------
    df : DataFrame with DATE and QTY_* columns
    brand_name : e.g., "B1", "B2", ...
    date_col : name of the datetime column (default "DATE")
    top_n : show only the top-N items by total sales for readability (None = show all)
    rotate_xticks : degrees to rotate x tick labels
    """

    dates = pd.to_datetime(df.index, errors="coerce")
        
    brand_prefix = f"QTY_{brand_name}_"
    brand_cols = [c for c in df.columns if c.startswith(brand_prefix)]
    if not brand_cols:
        raise ValueError(f"No quantity columns found for brand '{brand_name}'. Expected prefix '{brand_prefix}'.")
    
    # Year extraction
    df_year = df.copy()
    df_year["__YEAR__"] = dates.year
    years = sorted(df_year["__YEAR__"].dropna().unique())
    
    # Choose items (optionally top-N by total sales over all years)
    totals = df_year[brand_cols].sum().sort_values(ascending=False)
    keep_cols = totals.index.tolist() if top_n is None else totals.head(top_n).index.tolist()
    
    # Item labels (take the trailing part after QTY_Bx_)
    item_labels = [col.split("_")[-1] for col in keep_cols]
    
    # Build stacked values: one array per year aligned to keep_cols
    stacked_vals = []
    for y in years:
        yearly_sum = df_year.loc[df_year["__YEAR__"] == y, keep_cols].sum().to_numpy()
        stacked_vals.append(yearly_sum)
    
    # Plot
    x = np.arange(len(keep_cols))
    fig_w = max(10, len(keep_cols) * 0.45)  # widen if many items
    fig, ax = plt.subplots(figsize=(fig_w, 6))
    
    bottom = np.zeros(len(keep_cols))
    for y, vals in zip(years, stacked_vals):
        ax.bar(x, vals, bottom=bottom, label=str(int(y)))
        bottom += vals
    
    ax.set_xticks(x)
    ax.set_xticklabels(item_labels, rotation=rotate_xticks, ha="right")
    ax.set_xlabel("Item")
    ax.set_ylabel("Sales quantity (units)")
    ax.set_title(f"Brand {brand_name} - Stacked-by-year ")
    ax.legend(title="Year", ncol=min(4, len(years)))
    ax.grid(True, axis="y", alpha=0.3)
    ax.margins(x=0.01)
    plt.tight_layout()
    plt.show()
````

## Promotion Impact Analysis

Let's find the impact of promotion on the items

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from typing import Optional



def promotion_impact_plot(
    df: pd.DataFrame,
    top_n: int = 20,
    brand_filter: Optional[str] = None,
    qty_prefix: str = "QTY_",
    promo_prefix: str = "PROMO_",
):
    # Find matching (qty, promo) pairs by suffix (e.g., "B1_1")
    qty_cols = [c for c in df.columns if c.startswith(qty_prefix)]
    promo_cols = [c for c in df.columns if c.startswith(promo_prefix)]
    
    def suffix(name, prefix):
        return name[len(prefix):]
    
    qty_map = {suffix(c, qty_prefix): c for c in qty_cols}
    promo_map = {suffix(c, promo_prefix): c for c in promo_cols}
    common_suffixes = sorted(set(qty_map.keys()) & set(promo_map.keys()))
    
    if brand_filter is not None:
        common_suffixes = [s for s in common_suffixes if s.split("_")[0] == brand_filter]
    
    stats = []
    for sfx in common_suffixes:
        q_col = qty_map[sfx]
        p_col = promo_map[sfx]
        
        p = df[p_col] == 1
        np_mask = df[p_col] == 0
        
        promo_days = int(p.sum())
        no_promo_days = int(np_mask.sum())
        
        promo_qty_sum = float(df.loc[p, q_col].sum())
        no_promo_qty_sum = float(df.loc[np_mask, q_col].sum())
        total_qty = float(df[q_col].sum())
        
        promo_avg = float(df.loc[p, q_col].mean()) if promo_days > 0 else 0.0
        no_promo_avg = float(df.loc[np_mask, q_col].mean()) if no_promo_days > 0 else 0.0
        overall_avg = float(df[q_col].mean())
        
        brand = sfx.split("_")[0]
        item_id = sfx.split("_")[1] if "_" in sfx else sfx
        
        stats.append({
            "item_key": sfx,
            "brand": brand,
            "item": item_id,
            "promo_days": promo_days,
            "no_promo_days": no_promo_days,
            "promo_qty_sum": promo_qty_sum,
            "no_promo_qty_sum": no_promo_qty_sum,
            "promo_avg": promo_avg,
            "no_promo_avg": no_promo_avg,
            "total_qty": total_qty,
            "overall_avg": overall_avg,
        })
    
    stats_df = pd.DataFrame(stats)
    if stats_df.empty:
        raise ValueError("No items found with matching QTY_ and PROMO_ columns (after filtering).")
    
    top_df = stats_df.sort_values("total_qty", ascending=False).head(top_n)
    
    labels = top_df["item_key"].tolist()
    base = top_df["no_promo_avg"].to_numpy()
    stack = top_df["promo_avg"].to_numpy()
    
    x = np.arange(len(labels))
    fig_w = max(10, len(labels) * 0.45)
    fig, ax = plt.subplots(figsize=(fig_w, 6))
    ax.bar(x, base, label="No promotion")
    ax.bar(x, stack, bottom=base, label="With promotion")
    
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=90, ha="right")
    brand_note = f" (Brand {brand_filter})" if brand_filter else ""
    ax.set_title(f"Average Daily Sales by Item — with vs. without Promotion{brand_note}\nTop {top_n} items by total sales")
    ax.set_xlabel("Item (Brand_Item)")
    ax.set_ylabel("Average daily sales (units)")
    ax.legend()
    ax.grid(True, axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()
    

# Run for all brands, top 20
promotion_impact_plot(df, top_n=20, brand_filter=None)
````

## Heatmap Recurring pattern

Do a heat to find recurring pattern in days or months

````python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


def plot_store_dow_heatmap(
    df: pd.DataFrame,
    agg: str = "sum",
    per_year: bool = False,
):
    dates = pd.to_datetime(df.index, errors="coerce")
    if dates.isna().any():
        raise ValueError(f"Some values in index could not be parsed as datetime.")
    
    total_series = df["Total_QTY_Store"]
    tmp = pd.DataFrame({"DATE": dates, "total": total_series}).dropna(subset=["DATE"])
    
    iso = tmp["DATE"].dt.isocalendar()
    tmp["year"] = iso["year"].astype(int)
    tmp["week"] = iso["week"].astype(int)
    tmp["dow"] = tmp["DATE"].dt.dayofweek  # Monday=0
    
    def _plot_one(df_part: pd.DataFrame, title_suffix: str = ""):
        if agg == "mean":
            table = df_part.pivot_table(index="dow", columns=["year", "week"], values="total", aggfunc="mean")
        else:
            table = df_part.pivot_table(index="dow", columns=["year", "week"], values="total", aggfunc="sum")
        table = table.reindex(index=range(0, 7))
        data = table.to_numpy()
        
        fig_w = max(10, data.shape[1] * 0.15)
        fig, ax = plt.subplots(figsize=(fig_w, 4), constrained_layout=True)
        im = ax.imshow(data, aspect="auto")
        cbar = plt.colorbar(im, ax=ax)
        cbar.set_label(f"{'Sum' if agg=='sum' else 'Average'} sales per cell")
        
        ax.set_yticks(range(0,7))
        ax.set_yticklabels(["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
        
        col_labels = [f"{y}-W{int(w):02d}" for (y, w) in table.columns]
        n_cols = len(col_labels)
        if n_cols > 0:
            step = max(1, n_cols // 20)
            ticks = np.arange(0, n_cols, step)
            ax.set_xticks(ticks)
            ax.set_xticklabels([col_labels[i] for i in ticks], rotation=90, ha="right")
        ax.set_xlabel("ISO Week")
        ax.set_ylabel("Day of Week")
        ax.set_title(f"Store Total Sales — Day-of-Week Heatmap{title_suffix}")
        plt.show()
    
    if per_year:
        for yr in sorted(tmp["year"].unique()):
            _plot_one(tmp[tmp["year"] == yr], title_suffix=f" (Year {yr})")
    else:
        _plot_one(tmp, title_suffix="")

# Plot the heatmap
plot_store_dow_heatmap(df, agg="sum", per_year=True)
````

## How Ridge Regression Works in scikit-learn

### Overview

Ridge regression is a linear regression model with **L2 regularization**. It minimizes the following objective:

$$
[
\min_{\beta_0,\\ \beta}\\ \\frac{1}{n}\sum_{i=1}^n\big(y_i-\beta_0-\mathbf{x}_i^\top\beta\big)^2 \\;+\\; \alpha\\ \\|\beta|_2^2
]
$$

- $(\beta)$ has **one component per feature column** in `X` (your lags, rolling stats, one-hot dummies, month sin/cos, etc.).  
- $(\beta_0)$ is the intercept (**not** penalized).  
- $(\alpha)$ controls shrinkage strength. Larger $(\alpha)$ → coefficients are pulled closer to 0 (but rarely exactly 0; that would be Lasso).  

Under the hood, scikit-learn solves:

$$
[
(X^\\top X + \\alpha I)\\beta = X^\\top y
]
$$
---

### Where the Weights Are Assigned

In your feature matrix, Ridge assigns a coefficient to **each column**:

- `lag_1`, `lag_7`, `lag_14`, `lag_28`  
- `roll_mean_7`, `roll_mean_28`, `roll_std_7`, `roll_std_28`  
- `dow_0`, …, `dow_6` (day-of-week one-hot encoded)  
- `month_sin`, `month_cos`  

So yes—**weights are assigned to each of the columns in the feature vector**. There is no shared weight per lag size; it is strictly per feature.

---

### How Learning Happens with `pipe_manual.fit`

1. **Standardization**  
   Each feature column is standardized by `StandardScaler` (zero mean, unit variance). This ensures the Ridge penalty treats all features fairly.

2. **Ridge optimization**  
   The Ridge model learns coefficients by minimizing squared error + penalty term. It balances fitting the data with keeping weights small.

3. **Prediction**  
   At prediction time, input features are first scaled, then multiplied by the learned coefficients plus the intercept.

---

### Notes on Collinearity

If you include all 7 `dow_*` dummy variables **and** an intercept, they are collinear (sum to a vector of ones).  
- Ordinary least squares would complain.  
- Ridge regression is well-posed and can handle this.  
- For interpretability, it is common to drop one dummy or the intercept, though predictive performance usually changes little.

---

### Practical Notes for Time Series

- **Lag features**: You already avoided leakage by using shifted versions of `y`.  
- **Rolling statistics**: Also shifted to prevent peeking into the future.  
- **Temporal splits**: Keep using non-shuffled train/valid/test splits.  
- **Scaling**: Applied only to `X`, not `y`.  
- **Hyperparameter tuning**: Choosing `alpha` via validation is often more important than adding many more features.

---

### Key Takeaway

Ridge regression in this setup learns one weight per feature column (plus an intercept). These weights tell you how much each lag, rolling statistic, or seasonal indicator contributes to the forecast. Larger `alpha` shrinks them toward zero, improving generalization but reducing variance.

````python
# ----------------------------------------------------------------------
# 2) Feature engineering: lags, rolling stats, simple calendar encodings
# ----------------------------------------------------------------------
def make_features(y: pd.Series,
                  lags: List[int] = [1, 7, 14, 28],
                  roll_windows: List[int] = [7, 28],
                  add_calendar: bool = True):
    """Create a feature matrix X (df_feat) aligned with target y_out from a univariate daily series.
    Parameters
    ----------
    y : pd.Series
        The target time series on which to forecast (in this case, ts)
    lags : List[int], optional
        List of lagged features to create (default is [1, 7, 14, 28]).
    roll_windows : List[int], optional
        List of rolling window statistics to create (default is [7, 28], rolling window sizes).
    add_calendar : bool, optional
        Whether to add calendar features (default is True).
    ----------
    """
    df_feat = pd.DataFrame({"y": y})

    # Lag features
    for L in lags:
        df_feat[f"lag_{L}"] = y.shift(L)  # creates lag features (will also shorten index to prevent backfilling with -ve values)
        # lag features are simply the target variable shifted by L days
        # it is the 'recent history' that we want to look at

    # Rolling features (shifted by 1 to avoid leakage)
    for W in roll_windows:
        df_feat[f"roll_mean_{W}"] = y.shift(1).rolling(window=W, min_periods=1).mean()
        df_feat[f"roll_std_{W}"]  = y.shift(1).rolling(window=W, min_periods=1).std().fillna(0.0)

    if add_calendar:
        idx = y.index
        # Day-of-week (one-hot encoding)
        dow = pd.get_dummies(idx.dayofweek, prefix="dow")
        dow.index = idx
        df_feat = pd.concat([df_feat, dow], axis=1)
        # Month cyclic encoding 
        # Encode months on a cycle so that december is close to January, ensuring seasonal patterns
        month = idx.month
        df_feat["month_sin"] = np.sin(2 * np.pi * month / 12.0)
        df_feat["month_cos"] = np.cos(2 * np.pi * month / 12.0)
        # These help understand weekly, monthly seasonal patterns

    # Drop rows with NaNs from shifting (prevent out of bounds problems)
    # prevent -ve or NaN values from backfilling due to shifting in Lag features
    df_feat = df_feat.dropna()
    # Hence will shorten the dataframe index by 28 days (max Lag length)

    y_out = df_feat["y"].copy()
    X = df_feat.drop(columns=["y"])
    return X, y_out

X_all, y_all = make_features(ts)
````

## Explained: Lags, Rolling Stats, and Calendar Encodings for Time-Series Forecasting

This note explains the three feature families used in the univariate forecasting code and shows small numeric examples.

---

### 1) Lags (past values as features)

**Definition.** A lag **L** at time *t* is the target value *L* steps in the past:

$$\mathrm{lag}_L(t) = y_{t-L}.$$

**Why?** Lets the model learn autoregressive structure: today tends to resemble recent days.

**Example.** Daily sales over 8 days (t = 1..8):

$$y = [100,\,120,\,110,\,130,\,125,\,140,\,135,\,150].$$

At day *t = 8*:
- $lag_1(8) = y_7 = 135$
- $lag_7(8) = y_1 = 100$

---

### 2) Rolling Statistics (summaries of recent history)

We compute statistics over a trailing window and **shift by 1** so we only use the past (no leakage of \(y_t\)).

**Rolling mean** over window *W*:

$$\mathrm{roll\_mean}_W(t) = \frac{1}{W}\sum_{i=1}^{W} y_{t-i}.$$

**Rolling std** (sample standard deviation) over window *W*:

$$\mathrm{roll\_std}_W(t) = \sqrt{\frac{1}{W-1}\sum_{i=1}^{W}\big(y_{t-i}-\mathrm{roll\_mean}_W(t)\big)^2}.$$

**Why?** Captures recent level and volatility (trend/seasonality strength).

**Example (W = 3, t = 8).** Use days 7, 6, 5 → `[135, 140, 125]`.
- `roll_mean_3(8) = (135 + 140 + 125) / 3 = 133.33`
- `roll_std_3(8) ≈ 7.60`

---

### 3) Simple Calendar Encodings (periodic effects)

These are fixed, known features that inject weekly/annual seasonality.

#### 3.1 Day-of-Week One-Hot
For Monday = 0, …, Sunday = 6:

$$
\mathrm{dow}_k(t)=
\begin{cases}
1 & \text{if } \mathrm{day\_of\_week}(t)=k,\\
0 & \text{otherwise.}
\end{cases}
$$

If day 8 is **Wednesday (k = 2)**, then `[dow_0,…,dow_6] = [0,0,1,0,0,0,0]`.

#### 3.2 Month Cyclic Sine/Cosine
Encode month number $ (m_t \in \{1,\dots,12\} ) $ on a circle so December (12) is close to January (1):

$$
\mathrm{month\_sin}(t)=\sin\!\Big(\tfrac{2\pi m_t}{12}\Big), \qquad
\mathrm{month\_cos}(t)=\cos\!\Big(\tfrac{2\pi m_t}{12}\Big).
$$

**Examples:**
- January (m = 1): `sin = 0.5`, `cos ≈ 0.866`
- July (m = 7): `sin = −0.5`, `cos ≈ −0.866`

---

### Why these help (at a glance)
- **Lags:** expose short-term persistence (AR behavior).
- **Rolling stats:** summarize recent level and volatility without peeking ahead.
- **Calendar encodings:** inject known weekly/annual cycles that the model can exploit.

---

### Minimal pseudocode (how features are built)
```text
for each day t:
  features[t] = {
    lag_1 = y[t-1], lag_7 = y[t-7], lag_14 = y[t-14], ...
    roll_mean_7 = mean(y[t-1..t-7]), roll_std_7 = std(y[t-1..t-7]), ...
    dow_one_hot = one_hot(day_of_week(t))
    month_sin = sin(2π * month(t) / 12), month_cos = cos(2π * month(t) / 12)
  }

### Explained: Time-Aware Train/Validation/Test Split for Time Series

When evaluating a forecasting model, we must **respect time order** so the model never “sees the future.”  
A **time-aware split** uses three **contiguous** blocks from oldest → newest:

- **Train** — where the model learns patterns.
- **Validation** — the next block used for hyperparameter tuning/model selection.
- **Test** — the final (most recent) block for a one-time, fair evaluation.

> Unlike random splits, **no shuffling** occurs, and later blocks are strictly **after** earlier ones.

---

#### A Simple Example

Suppose your daily series runs **2017-01-01 → 2019-12-31**, and you choose:
- **Validation horizon** = 180 days  
- **Test horizon** = 180 days


```
[-------------------- TRAIN --------------------][----- VALID (180d) -----][----- TEST (180d) -----]
2017-01-01                                        ~2019-01 mid-2019         ~2019-12-31
```


In code (with `X_all`, `y_all` already built in time order):

```python
test_horizon  = 180
valid_horizon = 180

n = len(y_all)
test_start  = n - test_horizon
valid_start = test_start - valid_horizon

X_train, y_train = X_all.iloc[:valid_start], y_all.iloc[:valid_start]
X_valid, y_valid = X_all.iloc[valid_start:test_start], y_all.iloc[valid_start:test_start]
X_test,  y_test  = X_all.iloc[test_start:], y_all.iloc[test_start:]
```


#### Why This Matters
* Prevents leakage (no using future points to predict the past).

* Mirrors real-world forecasting: you always predict forward in time.

## RMSE and sMAPE for Store Sales Forecasting

This note explains two evaluation metrics used in our univariate Total_QTY_Store model and shows a tiny, concrete example.

---

### RMSE — Root Mean Squared Error
**What it measures:** Average size of the error in **units sold**, with **large mistakes penalized more** (squared).

$$

\mathrm{RMSE} = \sqrt{\frac{1}{n} \sum_{t=1}^{n} (y_t - \hat{y}_t)^2}

$$
**Tiny example (3 days)**  
Actual $(y)$: `[100, 120, 80]`  
Predicted $(\hat{y})$: `[110, 100, 90]`

Squared errors: 
- Day 1: $((110-100)^2 = 100)$  
- Day 2: $((100-120)^2 = 400)$  
- Day 3: $((90-80)^2 = 100)$

Mean squared error = $((100 + 400 + 100)/3 = 200)$  
**RMSE =** $(\sqrt{200} \approx 14.14)$ **units**

**Interpretation:** On average, predictions are ~14 units off, with big errors costing more.

---

### sMAPE — Symmetric Mean Absolute Percentage Error
**What it measures:** Average **relative** error as a **percentage**, scaled by the mean magnitude of actual and predicted. This makes errors more comparable between low and high sales levels.

**Formula**
$$
\mathrm{sMAPE} = \frac{100\%}{n} \sum_{t=1}^{n} \frac{\lvert y_t - \hat{y}_t\rvert}{\frac{\lvert y_t\rvert + \lvert \hat{y}_t\rvert}{2}} 
= \frac{100\%}{n} \sum_{t=1}^{n} \frac{2\lvert y_t - \hat{y}_t\rvert}{\lvert y_t\rvert + \lvert \\hat{y}_t\rvert}
$$
(In code we add a tiny `eps` in the denominator to avoid divide-by-zero when both are 0.)

**Same example (3 days)**
- Day 1: $( \frac{2\,|110-100|}{100+110} = \frac{20}{210} = 0.09524 \Rightarrow 9.52\% )$
- Day 2: $( \frac{2\,|100-120|}{120+100} = \frac{40}{220} = 0.18182 \Rightarrow 18.18\% )$
- Day 3: $( \frac{2\,|90-80|}{80+90} = \frac{20}{170} = 0.11765 \Rightarrow 11.76\% )$

**Average sMAPE ≈ (9.52% + 18.18% + 11.76%)/3 = 13.15%**

**Interpretation:** On average, predictions are off by ~13% of the typical (mean) size of that day’s actual and predicted sales.

---

#### When to use which?
- **RMSE:** use when you care about **absolute units** and want to penalize large mistakes more.
- **sMAPE:** use for **percentage comparability** across low/high sales periods (but it can be unstable when values are near zero; `eps` helps).


### MASE — Mean Absolute Scaled Error

**What it measures:** Model’s average absolute error **scaled** by a simple **seasonal naive** benchmark computed *in-sample*.  
- Unitless and comparable across series.  
- **MASE < 1** → the model is, on average, **better** than the seasonal naive baseline.  
- **MASE > 1** → worse than seasonal naive.

**Formula (with seasonal period \(m\))**
$$
\text{MASE} \;=\; \frac{\frac{1}{n}\sum_{t=1}^{n} |y_t - \hat{y}_t|}
{\frac{1}{N-m}\sum_{t=m+1}^{N} |Y_t - Y_{t-m}|}
$$
- Numerator: MAE on the **evaluation split** (e.g., test).  
- Denominator: in-sample **seasonal naive MAE** using lag \(m\) on the *training data used to fit the model* (often TRAIN+VALID). For non-seasonal data, set \(m=1\).

#### Tiny example (store total sales)

- **In-sample (TRAIN+VALID)** daily totals $(Y)$ for 8 days (assume weekly seasonality $(m=7)$):  
  $(Y = [100, 120, 110, 130, 125, 140, 135, 150]) $
  Seasonal naive differences (with $(m=7)$) use pairs $((Y_8, Y_1))$ only here:  
  $(|150 - 100| = 50)$ → in-sample seasonal naive MAE $(= 50)$.

- **Test (2 days)** actual vs predicted:  
  Actual $(y = [160, 155])$, Predicted $(\hat{y} = [150, 150])$  
  Test MAE $(= (|160-150| + |155-150|)/2 = (10 + 5)/2 = 7.5)$.

- **MASE** $(= \dfrac{7.5}{50} = 0.15)$  
  Interpretation: the model’s average absolute error is **15%** of the in-sample seasonal naive error → **better than naive**.


### Seasonal Naïve Benchmark & Its Role in MASE

#### What is the seasonal naïve forecast?
The **seasonal naïve** method predicts the next value by **repeating the value from one full season ago**.

- If your data are daily with weekly seasonality, the season length is $(m=7)$.
- The seasonal naïve forecast at time $(t)$ is:
$$
\hat{y}_t^{(\text{s-naive})} = y_{t-m}.
$$

This serves as a simple **baseline**: if a model can’t beat “repeat last week,” it may not be useful.

---

#### How it appears in the MASE formula
**MASE** scales your model’s MAE by the **in-sample** MAE of the seasonal naïve benchmark:

$$
\text{MASE} \;=\; 
\frac{\frac{1}{n}\sum_{t=1}^{n} \lvert y_t - \hat{y}_t \rvert}
{\underbrace{\frac{1}{N-m}\sum_{t=m+1}^{N} \lvert Y_t - Y_{t-m} \rvert}_{\text{in-sample seasonal naïve MAE}}}
$$

- $(Y)$ is the **in-sample** (training) series of length $(N)$.
- $(m)$ is the seasonal period (e.g., $(m=7)$ for weekly seasonality).
- **Numerator**: your model’s MAE on the evaluation split (e.g., test).
- **Denominator**: the average absolute error you’d get **in-sample** if you used the seasonal naïve rule $(\hat{Y}_t = Y_{t-m})$.

> Special case: **non-seasonal naïve** is $(m=1)$ (i.e., “tomorrow = today”).

---

#### Tiny example (weekly seasonality, $(m=7)$)
**In-sample** daily totals:
$$
Y = [100,\,120,\,110,\,130,\,125,\,140,\,135,\,150,\,145,\,160]
$$
Here $(N=10)$. Seasonal-naïve errors start at $(t=8)$ (first time we have $(t-7)$):

- $(t=8:\ |Y_8 - Y_1| = |150 - 100| = 50)$  
- $(t=9:\ |Y_9 - Y_2| = |145 - 120| = 25)$  
- $(t=10:\ |Y_{10} - Y_3| = |160 - 110| = 50)$

**In-sample seasonal naïve MAE** (MASE denominator):
$$
\frac{50 + 25 + 50}{3} = \frac{125}{3} \approx 41.67.
$$

**Test** actuals and predictions:
- Actual $(y = [160,\,155]), Predicted (\hat{y} = [150,\,150])$
- Test **MAE** (MASE numerator):
$$
\frac{|160-150| + |155-150|}{2} = \frac{10 + 5}{2} = 7.5.
$$

**MASE**:
$$
\text{MASE} = \frac{7.5}{41.67} \approx 0.18 \quad(< 1 \Rightarrow \text{better than seasonal naïve}).
$$

---

#### Takeaways
- **Seasonal naïve** = repeat last season’s value.
- **MASE < 1** ⇒ your model beats seasonal naïve on average.
- **MASE > 1** ⇒ your model underperforms seasonal naïve.


## ML Model: Ridge Regression with TimeSeriesSplit and GridSearch on Alpha

## ML Math: Ridge Regression

**α** (Alpha): What We’re Tuning

**Alpha** is the **L2-regularization strength** in Ridge regression — a knob that controls how much we shrink the model’s coefficients to combat overfitting.

---

### Ridge Regression: Objective Function (and role of Alpha)

Ridge finds weights $(w)$ by minimizing:
$$
\underbrace{\sum_t (y_t - \hat{y}_t)^2}_{\text{fit error}}
\;+\;
\alpha \,\underbrace{\|w\|_2^2}_{\text{L2 penalty}}.
$$

- **$(\alpha = 0)$** → Ordinary Least Squares (no shrinkage; highest variance).
- **Larger $(\alpha)$** → Stronger shrinkage; coefficients move toward 0 (lower variance, higher bias).
- In scikit-learn’s `Ridge`, the **intercept is not penalized**; only feature weights are.

---

### Why Grid-Search Alpha?

We don’t know beforehand how much shrinkage gives the best **out-of-sample** performance.  
So we try a set of values (e.g., `0.1, 1, 3, 10, 30, 100`) and use **time-aware cross-validation** to pick the $(\alpha)$ that forecasts best on later, unseen periods.

---

### Why It Matters for Our Features

Our design matrix includes **lags, rolling stats, and calendar encodings**, which can be correlated. L2 regularization:

- **Stabilizes** estimates when features are collinear (common with many lags).
- **Reduces overfitting** to noise/idiosyncrasies.
- **Shrinks** weights but generally does **not** set them exactly to zero (that’s more Lasso’s behavior).

---

### Tiny Intuition Example

Suppose `lag_1` and `lag_7` are highly correlated.

- **Small $(\alpha)$** might fit training very tightly with large, wobbly weights (e.g., $(w_1=2.1,\ w_7=-1.9$)) → poor generalization.
- **Moderate $(\alpha$)** shrinks both (e.g., $(w_1=0.9,\ w_7=0.6)$) → smoother predictor that generalizes better.

---

### Practical Tips

- Search $(\alpha)$ on a **log scale** (e.g., 0.1 → 100).
- **Standardize** features (as we do in the pipeline) so the penalty treats them comparably.
- If you also want **feature selection**, compare Ridge with **Lasso/ElasticNet**.

### `TimeSeriesSplit(n_splits=4)` — What It Does

**Short correction:** the argument is `n_splits` (plural).

`TimeSeriesSplit(n_splits=4)` creates **4 forward-in-time cross-validation folds** that **preserve time order**.  
Each fold trains on **earlier** observations and validates on the **immediately following** block.  
The training window **expands** with each fold (no shuffling).

**Why Use It?**
- Avoids **data leakage** (no peeking into the future).
- Mimics real forecasting: train on past → validate on later period.
- Enables reliable hyperparameter tuning (e.g., Ridge’s `alpha`).

**Tiny Example (10 time steps: indices 0..9)**
With `n_splits=4`, sklearn will split the series into 5 contiguous blocks of roughly equal size; each fold uses one block for validation and all prior blocks for training.

- **Fold 1**: Train = `[0, 1]` → Valid = `[2, 3]`  
- **Fold 2**: Train = `[0, 1, 2, 3]` → Valid = `[4, 5]`  
- **Fold 3**: Train = `[0, 1, 2, 3, 4, 5]` → Valid = `[6, 7]`  
- **Fold 4**: Train = `[0, 1, 2, 3, 4, 5, 6, 7]` → Valid = `[8, 9]`

Each evaluation moves **forward in time**, and the training set grows.

**Notes**
- In recent scikit-learn versions you can also control things like `test_size`, `gap`, or `max_train_size` (version-dependent) to fine-tune the validation block length, insert a gap between train/valid, or cap the train window.

## Question: Window of Scoring in CV?

How GridSearchCV Scores With `TimeSeriesSplit` — The Essentials

### 1) Time-aware splits
`TimeSeriesSplit(n_splits=4)` partitions the series in order into **5 contiguous blocks**. For each fold `$(k)$:

- **Train:** all blocks up to $(k)$  
- **Validate:** the **next** block (this is the **validation window**)

---

### 2) Score per fold (validation window only)
For fold $(k)$, compute the error **over the entire validation block**:

$$
\text{RMSE}_k \;=\; \sqrt{\frac{1}{n_k}\sum_{t \in \mathrm{val}_k}\big(y_t - \hat{y}_t\big)^2}
$$

Scikit-learn reports **`neg_root_mean_squared_error`** $(= -\,\text{RMSE}_k)$ so that **higher is better**.

---

### 3) Aggregate across folds
The cross-validated score is the **mean of fold scores**:

$$
\text{CV score} \;=\; \frac{1}{K}\sum_{k=1}^{K}\big(-\,\text{RMSE}_k\big)
$$

The hyperparameter with the **highest** average score (i.e., **lowest** average RMSE) wins.

---

### Tiny timeline example (10 time steps: indices 0..9)

With `n_splits=4`, sklearn creates 5 contiguous blocks of roughly equal size.  
Each fold uses one block for **validation** and all **prior** blocks for **training**:

- **Fold 1:** Train = `[0,1]` → **Validate = `[2,3]`**  
- **Fold 2:** Train = `[0,1,2,3]` → **Validate = `[4,5]`**  
- **Fold 3:** Train = `[0,1,2,3,4,5]` → **Validate = `[6,7]`**  
- **Fold 4:** Train = `[0,1,2,3,4,5,6,7]` → **Validate = `[8,9]`**

Each fold’s RMSE is computed **only** on its validation indices, then averaged.

---

** Notes & tips**
- The **training portion does not contribute** to the score—only the validation window does.  
- By default, folds are averaged **equally** (even if validation sizes differ). Some sklearn versions let you set `test_size`, `gap`, `max_train_size` to control fold sizes and add a gap to prevent leakage.  
- After selection, `best_estimator_` is typically **refit on TRAIN+VALID**; the final report is on the **held-out TEST** block you define separately.

````python
# ---------------------------------------------------
# 2) Exogenous features: promos + calendar encoding, one hot

# Will need to be modified for different Brand / items for Promotion Data specific to brand
# ---------------------------------------------------
def build_exog(frame: pd.DataFrame) -> pd.DataFrame:
    ex = pd.DataFrame(index=frame.index)

    # Promotions (same as before)
    promo_cols = [c for c in frame.columns if c.startswith("PROMO_")]
    if promo_cols:
        ex["promo_count"] = frame[promo_cols].sum(axis=1)
        ex["promo_share"] = ex["promo_count"] / len(promo_cols)
    else:
        ex["promo_count"] = 0.0
        ex["promo_share"] = 0.0

    # --- One-hot day-of-week (drop one to avoid collinearity with intercept) ---
    dow = frame.index.dayofweek  # Mon=0..Sun=6
    dow_ohe = pd.get_dummies(dow, prefix="dow", drop_first=True)
    dow_ohe.index = frame.index
    ex = pd.concat([ex, dow_ohe], axis=1)

    # --- One-hot month (optional; drop_one) ---
    mon = frame.index.month
    mon_ohe = pd.get_dummies(mon, prefix="mon", drop_first=True)
    mon_ohe.index = frame.index
    ex = pd.concat([ex, mon_ohe], axis=1)

    return ex.astype(float).fillna(0.0)
````

## Error Metrics

````python
# ---------------------------------------------------
# 3) Metrics: RMSE / sMAPE / MASE (m=7 for weekly)
# ---------------------------------------------------
def rmse(y_true, y_pred):
    return mean_squared_error(y_true, y_pred, squared=False)

def smape(y_true, y_pred, eps: float = 1e-8):
    y_true = np.asarray(y_true, float)
    y_pred = np.asarray(y_pred, float)
    return np.mean(2.0 * np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred) + eps)) * 100.0

def mase(y_true, y_pred, insample, m: int = 7, eps: float = 1e-8):
    insample = np.asarray(insample, float)
    if len(insample) <= m:
        return np.nan
    denom = np.mean(np.abs(insample[m:] - insample[:-m]))
    mae = np.mean(np.abs(np.asarray(y_true, float) - np.asarray(y_pred, float)))
    return mae / (denom + eps)
````

## ARIMAX & NARX: Stats Models

### ARIMAX

````python
# ---------------------------------------------------
# 4) Expanding-window ARIMAX (SARIMAX) backtest
# ---------------------------------------------------

# ---- Statsmodels for ARIMAX (SARIMAX) ----
from statsmodels.tsa.statespace.sarimax import SARIMAX

def expanding_forecast_sarimax(
    y: pd.Series,
    exog: pd.DataFrame,
    test_horizon: int = 180,
    order: Tuple[int,int,int] = (1,0,1),
    seasonal_order: Tuple[int,int,int,int] = (1,0,1,7),
    enforce_stationarity: bool = False,
    enforce_invertibility: bool = False,
) -> pd.Series:
    """
    1-step-ahead expanding window forecast using SARIMAX with exogenous regressors.
    """
    assert y.index.equals(exog.index), "y and exog must share the same index"

    # 1) Choose the first test index so the last `test_horizon` points are forecasts
    first_test_pos = len(y) - test_horizon
    preds = []
    idxs = y.index[first_test_pos:]

    # 2) For each test time t:
    for t in range(first_test_pos, len(y)):
        # Use data up to t-1 as the expanding training set
        y_train = y.iloc[:t]
        X_train = exog.iloc[:t]
        X_next  = exog.iloc[t:t+1]    # exog row for the next 1-step forecast

        # 3) Fit SARIMAX(y_train ~ exog_train, ARIMA/SARIMA errors)
        model = SARIMAX(endog=y_train, exog=X_train,
                        order=order, seasonal_order=seasonal_order,
                        enforce_stationarity=False, enforce_invertibility=False)
        res = model.fit(disp=True)

        # 4) Do a 1-step-ahead forecast at time t using exog at t
        fcast = res.forecast(steps=1, exog=X_next)
        preds.append(float(fcast.iloc[0]))

    # 5) Return predictions aligned with the test index
    return pd.Series(preds, index=idxs)
````

#### ARIMAX: Concept, Example, and Expanding-Window Forecasting

###### What ARIMAX Is
ARIMAX is a **dynamic regression** model: the target depends on **exogenous regressors** while the **residuals follow ARIMA/SARIMA** to capture remaining serial dependence:
$$
y_t = \beta_0 + \beta^\top x_t + \varepsilon_t, 
\qquad \varepsilon_t \sim \text{ARIMA}(p,d,q)\ \text{(or SARIMA with seasonality).}
$$

ARIMAX / sARIMAX will have to **Learn**  $\beta$ and ARIMA parameters through an optimization function

In retail, $(x_t)$ typically includes **promotions** and **calendar dummies** (day-of-week, month). The ARIMA part models autocorrelation/seasonality not explained by $(x_t)$.

###### Tiny Numerical Example
Let $(y_t)$ be total store sales with one regressor $(\text{promo}_t)$. Use an AR(1) error:
$$
y_t = \beta_0 + \beta_{\text{promo}}\,\text{promo}_t + \varepsilon_t, 
\qquad \varepsilon_t = \phi\,\varepsilon_{t-1} + u_t,\ \ u_t\ \text{i.i.d.}
$$
Take $(\beta_0=100)$, $(\beta_{\text{promo}}=20)$, $(\phi=0.6)$.  
If $(\text{promo}_t=1)$ and $(\varepsilon_{t-1}=10)$:
- Mean part: $(\mathbb{E}[y_t|x_t] = 100 + 20\cdot 1 = 120)$  
- Error persistence: $(\mathbb{E}[\varepsilon_t|\varepsilon_{t-1}] = 0.6\times 10 = 6)$

The fitted SARIMAX learns both $(\beta)$ and ARIMA/SARIMA parameters.

###### Expanding-Window Evaluation
An **expanding window** respects time order:
1. Fit on an initial span (past only), forecast the next point(s).
2. **Expand** the training set to include those past points.
3. Repeat through the test horizon.

Aggregate the rolling errors (e.g., RMSE, sMAPE, MASE). This mirrors real forecasting and avoids leakage.

###### How `expanding_forecast_sarimax()` Works
1. Pick the first test index so the last `test_horizon` days are forecasts.  
2. For each test time `t`:
   - Train on all data **up to t−1** (`y.iloc[:t]`, `exog.iloc[:t]`).
   - Fit **SARIMAX(endog=y_train, exog=X_train, order=(p,d,q), seasonal_order=(P,D,Q,s))**.
   - **Forecast 1-step ahead** at `t` using `exog.iloc[t]`; collect the prediction.
3. Return the predictions aligned with the test dates.

> Exogenous features (promotions, DOW/month dummies) shape the **mean**; the ARIMA/SARIMA component models **serial dependence/seasonality** in the **residuals**. This combination is the essence of dynamic regression.



##### What SARIMAX fits (the objective)
`statsmodels` **SARIMAX** estimates parameters by **Maximum Likelihood (ML)**. It expresses the model in a **state‑space** form, runs a **Kalman filter** to compute one‑step‑ahead prediction errors and their variances, sums their log‑densities to form the **log‑likelihood** $(\ell(\theta))$, and then **maximizes** $(\ell(\theta))$ (equivalently minimizes $(-\ell(\theta))$).

- **Observation (with exogenous regressors $(x_t)$)**:
  $$
  y_t = \beta^\top x_t + c + \varepsilon_t, \quad \varepsilon_t \sim \mathcal N(0,\sigma^2)
  $$
- **State/transition equations** encode the ARMA/SARIMA dynamics for the latent state that generates $(\varepsilon_t)$.

##### What optimizer is used?
By default, SARIMAX uses a **quasi‑Newton optimizer (L‑BFGS‑B)**. Other SciPy optimizers (e.g., BFGS, Newton, Powell, Nelder–Mead, Conjugate Gradient) can be selected, but the default is typically L‑BFGS‑B. The optimizer iteratively updates all parameters (AR/MA, seasonal terms, $(\beta)$, variance) to **maximize the likelihood**.

##### How does it “search” for different $(\beta)$?
It does **not** grid‑search $(\beta)$. The exogenous coefficients $(\beta)$ are estimated **jointly** with ARMA/SARIMA parameters by **continuous ML optimization**. The Kalman filter evaluates the likelihood for each trial parameter vector $(\theta)$; the optimizer then moves $(\theta)$ to improve $(\ell(\theta))$.

##### What does $(\varepsilon)$ represent?
After removing the regression mean (and any differencing), the **unexplained part** is modeled by an ARMA/SARIMA error:

- **Innovation** $(u_t)$: white‑noise shock, $(u_t \sim \mathcal N(0,\sigma^2))$.
- **Residual/error** $(\varepsilon_t)$: follows serial dynamics driven by innovations. For an AR(1):
  $$
  \varepsilon_t = \phi\,\varepsilon_{t-1} + u_t.
  $$

The Kalman filter works with **one‑step‑ahead forecast errors (innovations)** to build the likelihood.

##### What is the order $((p,d,q))$?
The nonseasonal **ARIMA** order is:
- $(p)$: AR order (dependence on past values/errors via autoregression),
- $(d)$: number of **differences** applied to $(y)$,
- $(q)$: MA order (dependence on past **innovations**).

With exogenous regressors (ARIMAX), the **mean** is
$$
\mu_t = c + \beta^\top x_t \quad (\text{after differencing if } d>0),
$$
while the deviations $(y_t - \mu_t)$ follow ARMA$((p,q))$. Seasonality extends this with $((P,D,Q,s))$.

##### Tiny 1‑D ARIMAX example (one input)
**Model** (ARIMAX(1,0,0) with one regressor $(x_t)$):
$$
\begin{aligned}
 y_t &= c + \beta x_t + \varepsilon_t,\\
 \varepsilon_t &= \phi\,\varepsilon_{t-1} + u_t, \quad u_t \sim \mathcal N(0,\sigma^2).
\end{aligned}
$$
**Numbers.** Let $(c=50)$, $(\beta=10)$, $(\phi=0.6)$. Suppose yesterday’s residual was $(\varepsilon_{t-1}=5)$; today $(x_t=1)$.
- **Mean part:** $(\mu_t = 50 + 10\cdot 1 = 60)$.
- **Error persistence (expected):** $(\mathbb E[\varepsilon_t\mid\varepsilon_{t-1}] = 0.6\times 5 = 3)$.
- **One‑step forecast (conditional mean of $(y_t)$):** $(\mathbb E[y_t] = \mu_t = 60)$. The realized value is $(y_t = 60 + \varepsilon_t)$.

**How SARIMAX fits this:** Given a history $(\{(y_t, x_t)\})$, the algorithm: (1) proposes parameters $(\theta=(c,\beta,\phi,\sigma^2))$; (2) runs the Kalman filter to get innovations and their variances → log‑likelihood; (3) updates $(\theta)$ via L‑BFGS‑B to increase the likelihood; (4) iterates to convergence.

##### How `statsmodels.tsa.statespace.SARIMAX` works (at a glance)
1. You specify `order=(p,d,q)`, `seasonal_order=(P,D,Q,s)`, and `exog`.
2. The model is cast to a **state‑space** system (observation + state equations).
3. The **Kalman filter** yields prediction errors and variances → **log‑likelihood**.
4. A SciPy optimizer (default **L‑BFGS‑B**) **maximizes** the likelihood over all parameters.
5. The **Kalman smoother** provides smoothed states; `forecast` / `get_forecast` use fitted params (and future exog) to generate predictions and intervals.

#### SARIMAX on 1‑D Sales: How (p,d,q) and (P,D,Q,s) Change the Forecast

Below is a compact, numeric example showing how different SARIMAX orders affect a **1‑step‑ahead** forecast for daily store sales.

---

###### Shared Setup (used in every case)
- Exogenous driver: a promotion flag $(x_t\in\{0,1\})$.
- Regression mean: $(\mu_t = c + \beta x_t)$ with $(c=50,\ \beta=10)$.
- **Today**: $(x_t=1 \Rightarrow \mu_t=60)$.
- **Yesterday**: $(x_{t-1}=0\Rightarrow \mu_{t-1}=50)$, observed $(y_{t-1}=54)$.
  - Residual yesterday: $(\varepsilon_{t-1}=y_{t-1}-\mu_{t-1}=54-50=4)$.
- **Last week** (7 days ago): $(\varepsilon_{t-7}=5)$ (from filtering history).
- **Last innovation** (from Kalman filter): $(u_{t-1}=-2)$.
- Coefficients used in examples:
  - Nonseasonal AR(1) coefficient $(\phi=0.5)$
  - Nonseasonal MA(1) coefficient $(\theta=0.4)$
  - Seasonal AR(1) coefficient (weekly) $(\Phi=0.6)$

> SARIMAX models the **mean** via exogenous regressors and the leftover **error** via ARMA/SARIMA. For a 1‑step forecast, the expected innovation is zero, so only **known past residuals/innovations** enter.

---

###### Case A — ARIMAX(1,0,0): AR(1) errors, no seasonality
Order $((p,d,q)=(1,0,0), (P,D,Q,s)=(0,0,0,7))$

Error model: $(\varepsilon_t = \phi\,\varepsilon_{t-1} + u_t)$.

One‑step forecast:
$$
\hat y_t = \mu_t + \phi\,\varepsilon_{t-1} = 60 + 0.5\times 4 = \mathbf{62}.
$$

---

###### Case B — ARIMAX(0,1,1): first difference + MA(1) errors, no seasonality
Order $((0,1,1))$

Work on differences $(\Delta y_t = y_t - y_{t-1})$:
$$
\Delta y_t = \mu_t + \theta\,u_{t-1} + u_t.
$$
1‑step forecast for the **change**:
$$
\widehat{\Delta y_t} = \mu_t + \theta\,u_{t-1} = 60 + 0.4\times(-2) = 59.2.
$$
Convert to a **level** forecast:
$$
\hat y_t = y_{t-1} + \widehat{\Delta y_t} = 54 + 59.2 = \mathbf{113.2}.
$$

> Differencing switches the model to predicting **increments**; level forecasts become yesterday’s level plus the predicted change.

---

###### Case C — SARIMAX with seasonal AR(1) only
Order $((0,0,0))$ with seasonal $((1,0,0,7))$

Error model: $(\varepsilon_t = \Phi\,\varepsilon_{t-7} + u_t)$.

One‑step forecast:
$$
\hat y_t = \mu_t + \Phi\,\varepsilon_{t-7} = 60 + 0.6\times 5 = \mathbf{63}.
$$

> Seasonal AR(1) “pulls” the forecast toward last week’s leftover pattern.

---

###### Case D — SARIMAX with both AR(1) and seasonal AR(1)
Order $((1,0,0))$ and seasonal $((1,0,0,7))$

Error model: $(\varepsilon_t = \phi\,\varepsilon_{t-1} + \Phi\,\varepsilon_{t-7} + u_t)$.

One‑step forecast:
$$
\hat y_t = \mu_t + \phi\,\varepsilon_{t-1} + \Phi\,\varepsilon_{t-7}
= 60 + 0.5\times 4 + 0.6\times 5 = 60 + 2 + 3 = \mathbf{65}.
$$

---

###### What changes when you vary the orders?
- **$(p)$** (AR order): adds dependence on **recent residuals** (e.g., $(\varepsilon_{t-1})$), nudging the forecast toward what just happened.
- **$(d)$** (differences): the model predicts **changes**; level forecasts are **yesterday + predicted change**.
- **$(q)$** (MA order): incorporates **past innovations** $(u_{t-1}, u_{t-2},\ldots)$ to adjust the next step.
- **Seasonal $((P,D,Q,s))$**: the same ideas but at **lag $(s)$** (e.g., $(s=7)$ for weekly). Seasonal AR uses $(\varepsilon_{t-s})$; seasonal differencing ($(D=1)$) targets **week‑over‑week changes**; seasonal MA uses **seasonal innovations**.

> In actual fitting, residuals $(\varepsilon_{t-k})$ and innovations $(u_{t-k})$ are computed by the Kalman filter. These formulas simply illustrate each order’s **structural effect** on a 1‑step forecast.

### NARX

````python
# ---------------------------------------------------
# 5) NARX-like (nonlinear) backtest via MLP on lagged y + exog lags
# ---------------------------------------------------
def make_narx_matrix(
    y: pd.Series,
    exog: pd.DataFrame,
    y_lags: Iterable[int] = (1, 7, 14, 28),
    x_lags: Iterable[int] = (0, 1, 7),
) -> Tuple[pd.DataFrame, pd.Series]:
    """
    Build feature matrix for NARX: past y (autoregressive lags) + exogenous (with lags).
    x_lags includes 0 to use contemporaneous exog_t (common in dynamic regression/NARX).
    """
    X = pd.DataFrame(index=y.index)
    # y lags
    for L in y_lags:
        X[f"y_lag_{L}"] = y.shift(L)
    # exog lags
    for col in exog.columns:
        for L in x_lags:
            X[f"{col}_lag_{L}"] = exog[col].shift(L)
    # align and drop NaNs from shifting
    Z = pd.concat([X, y.rename("y")], axis=1).dropna()
    return Z.drop(columns=["y"]), Z["y"]

def expanding_forecast_narx(
    y: pd.Series,
    exog: pd.DataFrame,
    test_horizon: int = 180,
    y_lags: Iterable[int] = (1, 7, 14, 28),
    x_lags: Iterable[int] = (0, 1, 7),
    mlp_kwargs: Optional[dict] = None,
) -> pd.Series:
    """
    1-step-ahead expanding window forecast using an MLP on NARX features.
    """
    X_all, y_all = make_narx_matrix(y, exog, y_lags=y_lags, x_lags=x_lags)

    # Determine first test index within the aligned (lagged) matrix
    full_index = X_all.index
    first_test_idx = full_index[-test_horizon]  # go by aligned design, not raw y

    if mlp_kwargs is None:
        mlp_kwargs = dict(
            hidden_layer_sizes=(64,),
            activation="relu",
            random_state=0,
            max_iter=500,
            learning_rate_init=0.001,
        )

    preds = []
    test_index = full_index[full_index >= first_test_idx]

    for t_idx in test_index:
        # expanding window: use rows strictly before t_idx
        train_mask = full_index < t_idx
        X_tr, y_tr = X_all.loc[train_mask], y_all.loc[train_mask]
        X_te = X_all.loc[[t_idx]]

        model = Pipeline([
            ("scaler", StandardScaler()),
            ("mlp", MLPRegressor(**mlp_kwargs))
        ])
        model.fit(X_tr, y_tr)
        yhat = model.predict(X_te)[0]
        preds.append(yhat)

    return pd.Series(preds, index=test_index)
````

#### NARX: explained

**NARX** stands for **Nonlinear AutoRegressive model with eXogenous inputs**. It predicts today’s value $(y_t)$ from:
- recent **lags of the series** $(y_{t-1},\ldots,y_{t-p})$, and
- current/past **external drivers** $(x_t, x_{t-1},\ldots,x_{t-q})$

through a (possibly) **nonlinear** function $(f(\cdot))$ plus noise:
$$
 y_t = f\big(\,y_{t-1:t-p},\; x_{t:t-q}\,\big) + \varepsilon_t.
$$
If $(f)$ is linear, this reduces to **ARX**. Using a neural net, tree, or any nonlinear regressor for $(f)$ makes it **NARX**.

---

###### How NARX Computes Predictions
For a **1‑step‑ahead** forecast at time $(t)$:
$$
 \hat y_t = f\big(\,y_{t-1:t-p},\; x_{t:t-q}\,\big).
$$
- You use **observed past** $(y)$ lags and the **known exogenous** values (often including $(x_t)$).
- For **multi‑step** forecasts, you feed previous **predictions** back in place of unknown future $(y)$’s (recursive forecasting). Exogenous futures must be known or assumed.

---

###### Where These Features Come From (conceptually)
To train $(f)$, we build a supervised dataset with columns like:
- `y_lag_L` for each $(y_{t-L})$ you include,
- exogenous lag columns for each $(x_{t-L})$ you include (often including $(L=0)$ so you can use $(x_t)$ at prediction time).

Training uses **past rows only** for a given forecast time to avoid leakage (e.g., with an **expanding window**: train on all dates before $(t)$, predict $(t)$, then roll forward).

---

###### Simple Numerical Example (Retail Sales)
Suppose we choose:
- lags $(y_{t-1})$ and $(y_{t-7})$,
- exogenous promos $(\text{promo}_t)$ and $(\text{promo}_{t-1})$,
- and a small **nonlinear** function (include a quadratic term on $(y_{t-1})$):
$$
\hat y_t = 5 + 0.6\,y_{t-1} + 0.2\,y_{t-7} + 8\,\text{promo}_t + 2\,\text{promo}_{t-1} + 0.001\,y_{t-1}^2.
$$
Plug in: $(y_{t-1}=120)$, $(y_{t-7}=100)$, $(\text{promo}_t=1)$, $(\text{promo}_{t-1}=0)$.

Compute terms:
- $(0.6\cdot120 = 72)$
- $(0.2\cdot100 = 20)$
- $(8\cdot1 = 8)$
- $(2\cdot0 = 0)$
- $(0.001\cdot(120)^2 = 0.001\cdot 14400 = 14.4)$
- Constant $(= 5)$

Sum: $(5 + 72 + 20 + 8 + 0 + 14.4 = \mathbf{119.4})$.

So the **NARX 1‑step forecast** is $(\hat y_t \approx 119.4)$ units. The quadratic term makes this **nonlinear**; a neural network would learn such effects from data automatically.

---

###### Why Use NARX?
- Captures **serial dependence** via $(y)$ lags **and** exogenous effects (e.g., promotions, calendar).
- Flexible nonlinear $(f)$ can model **thresholds, interactions, and saturation** that linear models miss.
- Works naturally with **expanding‑window** training to mirror real‑time forecasting.

````python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import numpy as np
import pandas as pd

def expanding_forecast_sarimax_robust(
    y: pd.Series,
    exog: pd.DataFrame,
    test_horizon: int = 180,
    order=(1,0,1),
    seasonal_order=(1,0,0,7),
    trend: str = "c",                   # use "n" if you keep ALL dummies
    enforce_stationarity: bool = True,
    enforce_invertibility: bool = True,
    simple_differencing: bool = True,
    concentrate_scale: bool = True,
    maxiter: int = 300,
    tol: float = 1e-6,
    methods = ("lbfgs", "powell"),     # try LBFGS, then Powell if needed
    warm_start: bool = True,
    refit_every: int = 7,              # refit once per week
    standardize_exog: bool = False     # optionally z-score exog for stability
) -> pd.Series:
    """
    1-step ahead expanding forecasts for SARIMAX with convergence-friendly settings.
    """

    assert y.index.equals(exog.index), "y and exog must share the same index"
    y = y.asfreq("D")
    exog = exog.asfreq("D").copy()

    # Optional: standardize exogenous regressors (can help optimizer)
    if standardize_exog:
        exog = exog.apply(lambda s: (s - s.mean()) / (s.std(ddof=0) or 1.0))

    first_test_pos = len(y) - test_horizon
    preds = []
    idxs = y.index[first_test_pos:]

    prev_params = None
    res = None  # last successful result object

    for i, t in enumerate(range(first_test_pos, len(y))):
        # Refit only every 'refit_every' steps (and at the beginning)
        need_refit = (i % refit_every == 0) or (res is None)

        y_train = y.iloc[:t]
        X_train = exog.iloc[:t]
        X_next  = exog.iloc[t:t+1]

        if need_refit:
            model = SARIMAX(
                endog=y_train,
                exog=X_train,
                order=order,
                seasonal_order=seasonal_order,
                trend=trend,
                enforce_stationarity=enforce_stationarity,
                enforce_invertibility=enforce_invertibility,
                simple_differencing=simple_differencing,
                concentrate_scale=concentrate_scale,
            )

            # Attempt: LBFGS -> Powell fallback
            fit_ok = False
            start_params = (prev_params if (warm_start and prev_params is not None) else None)
            for m in methods:
                try:
                    res = model.fit(
                        method=m,
                        maxiter=maxiter,
                        disp=False,
                        tol=tol,
                        start_params=start_params
                    )
                    fit_ok = True
                    break
                except Exception:
                    # try next method
                    start_params = None  # reset for more robust fallback
                    continue

            if not fit_ok:
                # Last resort: relax constraints and try again
                model = SARIMAX(
                    endog=y_train,
                    exog=X_train,
                    order=order,
                    seasonal_order=seasonal_order,
                    trend=trend,
                    enforce_stationarity=False,
                    enforce_invertibility=False,
                    simple_differencing=simple_differencing,
                    concentrate_scale=concentrate_scale,
                )
                res = model.fit(method="powell", maxiter=maxiter, disp=False)

            prev_params = res.params

        # 1-step ahead forecast at time t
        fcast = res.forecast(steps=1, exog=X_next)
        preds.append(float(fcast.iloc[0]))

    return pd.Series(preds, index=idxs)
````

````python
# ---------------------------------------------------
# 3) Windowing utilities (w=30, hop=1)
#    Build sequences once, then index by target timestamps
# ---------------------------------------------------
def build_sequences(y_series: pd.Series,
                    X_frame: pd.DataFrame,
                    window: int = 30,
                    hop: int = 1):
    """
    Build rolling windows ending at each t (target is y[t]),
    returning tensors and the corresponding target timestamps.
    """
    y_vals = y_series.values.astype(float)
    X_vals = X_frame.values.astype(float)
    n = len(y_series)
    n_exog = X_vals.shape[1]

    y_seq_list = []
    X_seq_list = []
    y_target = []
    target_index = []

    # We need indices t such that t-window..t-1 exist for inputs, and target is y[t]
    t = window
    while t < n:
        if (t - window) >= 0:
            y_win = y_vals[t-window:t].reshape(window, 1)        # (w, 1)
            X_win = X_vals[t-window:t].reshape(window, n_exog)   # (w, n_exog)

            y_seq_list.append(y_win)
            X_seq_list.append(X_win)
            y_target.append(y_vals[t])
            target_index.append(y_series.index[t])
        t += hop

    y_seq = np.stack(y_seq_list, axis=0)               # (samples, w, 1)
    X_seq = np.stack(X_seq_list, axis=0)               # (samples, w, n_exog)
    y_target = np.array(y_target, dtype=float)         # (samples,)
    target_index = pd.to_datetime(pd.Index(target_index))
    return y_seq, X_seq, y_target, target_index
````

### Explained `build_sequences()`: What It Does (with a Tiny Numeric Example)

#### Purpose
`build_sequences(y, X, window=w, hop=h)` converts a time series and its aligned exogenous design into a **supervised, fixed‑length, sliding‑window** dataset suitable for models like **1D‑CNN (on `y`)** plus **MLP (on `X`)**.

#### Inputs
- **`y`**: univariate series (indexed by date).
- **`X`**: exogenous dataframe with `n_exog` columns (same dates as `y`).
- **`window = w`**: how many **past** days to use as inputs for each target.
- **`hop = h`**: how far to slide the window each step (usually `h=1`).

#### Outputs
For each valid target time $(t)$ (where the previous `w` days exist), it returns:
- **`y_seq`**: the last `w` values of `y` stacked as a sequence → shape **(N, w, 1)**
- **`X_seq`**: the last `w` rows of `X` for the same dates → shape **(N, w, n_exog)**
- **`y_target`**: the next‑day value $(y_t)$ → shape **(N, )**
- **`target_index`**: the timestamps for those targets → length **N**

Here, **N** is the number of windows produced when sliding across the series with stride `hop`.

#### How the windows are built
For each target time $(t)$:
- **y‑window**: $([y_{t-w},\dots,y_{t-1}])$ → used by the CNN branch
- **X‑window**: $([X_{t-w},\dots,X_{t-1}])$ → used by the MLP branch
- **target**: $(y_t)$ (the value to predict one step ahead)

### Scaling Fit (Training-Only Standardization): What It Does and What It Doesn’t

##### TL;DR
- **Scaling fit ≠ model fit.**
- `fit()` on the scalers computes **means/standard deviations** on the **in-sample period only**; the ML model is trained **after** this, on the **scaled** data.
- Using training-only statistics prevents **data leakage**.

---

##### What the scaling block does
1. **Fit scalers on in-sample (training) data only**
   - Target: compute $(\mu_y, \sigma_y)$ from 2014–2017.
   - Exogenous columns: for each feature $(X^{(j)})$, compute $(\mu_j, \sigma_j)$ from 2014–2017.
2. **Transform all dates** (train + test) with those fixed parameters
   - Target standardization: $[ z_y = \frac{y - \mu_y}{\sigma_y} ]$
   - Feature standardization (column-wise): $[ z_j = \frac{X^{(j)} - \mu_j}{\sigma_j} ]$
3. **Train the model on the scaled sequences** and
4. **Invert only the target** after prediction: $[ \hat y = \mu_y + \sigma_y\,\hat z_y. ]$

> Because the scaler’s `fit` uses **only** the in-sample window, the test period remains unseen during preprocessing —
> no leakage.

---

##### Why scale?
- Stabilizes and speeds up optimization (especially for neural nets).
- Puts heterogeneous exogenous features on comparable scales.
- Keeps evaluation honest by freezing scaling stats to the training period.

---

##### Tiny numerical example
**In-sample target values (2014–2017):** $([100,\ 120,\ 80])$

- Mean: $(\mu_y = (100+120+80)/3 = 100)$
- Std (population-like): $(\sigma_y = \sqrt{\tfrac{0^2+20^2+(-20)^2}{3}} = \sqrt{\tfrac{800}{3}} \approx 16.33)$

**Out-of-sample day (2018):** $(y=110)$

- Scale it: $(z_y = (110 - 100)/16.33 \approx 0.612)$
- Suppose the model predicts $(\hat z_y = 0.25)$. Invert:
  $[ \hat y = \mu_y + \sigma_y\,\hat z_y = 100 + 16.33\times 0.25 \approx 104.1. ]$

**Features** are handled the same column-wise with their own $(\mu_j,\sigma_j)$ computed from 2014–2017 only.

---

##### Sanity checklist
- ✅ Fit scalers on **training** slice only.
- ✅ Build windows from the **already-scaled** series.
- ✅ Inverse-transform **only** the target.
- ❌ Do **not** refit scalers on the entire dataset (would leak test info).

---

##### Bottom line
The “scaling fit” stage only learns **preprocessing parameters** (means/stds). The actual ML model (e.g., CNN+MLP) is
trained **afterward** on the scaled data; scaling fit does **not** count as fitting the predictive model itself.

## CNN + MLP: Branch merge

````python
# ---------------------------------------------------
# 5) Define the dual-branch CNN+MLP model
# ---------------------------------------------------
def build_cnn_mlp(window: int, n_exog: int) -> Model:
    # Branch A: 1D CNN over y sequence
    inp_y = layers.Input(shape=(window, 1), name="y_seq")
    a = layers.Conv1D(16, 3, padding="causal", activation="relu")(inp_y)  # 16 output filters, 3 kernel size (1D)
    a = layers.Conv1D(16, 3, padding="causal", activation="relu")(a)
    a = layers.GlobalAveragePooling1D()(a)

    # Branch B: MLP over exogenous window (flattened)
    inp_x = layers.Input(shape=(window, n_exog), name="x_seq")
    b = layers.Flatten()(inp_x)
    b = layers.Dense(64, activation="relu")(b)

    # Fuse the layers
    h = layers.Concatenate()([a, b])
    h = layers.Dense(64, activation="relu")(h)
    out = layers.Dense(1, name="y_hat")(h)

    model = Model([inp_y, inp_x], out)
    model.compile(optimizer="adam", loss="mse")
    return model
````

## Train defn: Expanding Window

````python
# ---------------------------------------------------
# 6) Expanding-window forecasting on 2018 (refit every k days)
# ---------------------------------------------------
def expanding_forecast_cnn_mlp(
    y_seq_all, X_seq_all, y_tgt_all, tgt_idx_all,
    train_end_date: pd.Timestamp,
    test_start_date: pd.Timestamp,
    test_end_date: pd.Timestamp,
    y_scaler: StandardScaler,
    refit_every: int = 7,
    epochs_per_refit: int = 10,
    batch_size: int = 64,
    val_fraction: float = 0.1,
    verbose: int = 0,
):
    """
    1-step-ahead expanding-window forecasting:
    - Train on all windows with target_index <= train_end_date initially
    - For t in [test_start_date, test_end_date]:
        * Refit model every `refit_every` days using all up-to-(t-1) windows
        * Predict 1-step ahead for date t (using the window ending at t-1)
    """

    # Identify initial training windows and the test target indices
    initial_train_mask = (tgt_idx_all <= train_end_date)
    test_mask = (tgt_idx_all >= test_start_date) & (tgt_idx_all <= test_end_date)

    if initial_train_mask.sum() == 0 or test_mask.sum() == 0:
        raise ValueError("No training or test windows found for the specified dates.")

    # Prepare containers
    preds = []
    pred_idx = tgt_idx_all[test_mask]
    n_exog = X_seq_all.shape[-1]

    # Model
    tf.random.set_seed(42)
    model = build_cnn_mlp(window=y_seq_all.shape[1], n_exog=n_exog)

    # Helper: fit model on windows up to a cutoff date
    def fit_up_to(date_cutoff):
        train_mask = (tgt_idx_all <= date_cutoff)
        Xy = np.arange(len(tgt_idx_all))[train_mask]
        if len(Xy) < 100:
            # very small training set — reduce epochs
            local_epochs = max(3, epochs_per_refit // 2)
        else:
            local_epochs = epochs_per_refit

        # simple time-based split: last val_fraction as validation
        n_train = len(Xy)
        n_val = int(max(1, np.floor(val_fraction * n_train)))
        X_train_idx = Xy[: n_train - n_val]
        X_val_idx   = Xy[n_train - n_val :]

        history = model.fit(
            [y_seq_all[X_train_idx], X_seq_all[X_train_idx]],
            y_tgt_all[X_train_idx],
            validation_data=([y_seq_all[X_val_idx], X_seq_all[X_val_idx]], y_tgt_all[X_val_idx]),
            epochs=local_epochs,
            batch_size=batch_size,
            verbose=verbose
        )
        return history

    # Initial fit on in-sample (<= train_end_date)
    fit_up_to(train_end_date)

    # Walk through test dates, refitting periodically
    last_refit_date = train_end_date
    for i, t in enumerate(pred_idx):
        # refit every k days using data up to t-1
        if (i % refit_every == 0) and (t > last_refit_date):
            fit_up_to(t - pd.Timedelta(days=1))
            last_refit_date = t

        # find the input window that ends at t-1 → its target is t
        # (in our construction, the sample whose target_index == t holds the correct window)
        row = int(tgt_lut.loc[t])
        yhat_scaled = model.predict([y_seq_all[row:row+1], X_seq_all[row:row+1]], verbose=0)[0, 0]
        # invert scaling
        yhat = y_scaler.inverse_transform([[yhat_scaled]])[0, 0]
        preds.append(yhat)

    return pd.Series(preds, index=pred_idx, name="CNN_MLP_pred")
````

### `expanding_forecast_cnn_mlp(...)`: What It Does and How the Numbers Line Up

#### Goal
Produce **1‑step‑ahead forecasts** on the test period (e.g., 2018) using a dual‑branch model:
- **1D‑CNN** on the last `w` target values (the y‑sequence), and
- **MLP** on the last `w` rows of exogenous features (the X‑sequence),

while training in a **true expanding‑window** fashion (only past data up to “yesterday” are used at each step).

---

#### Inputs (already prepared earlier)
- **`y_seq_all`** ∈ ℝ^{N×w×1}: sliding windows of the **scaled** target (last `w` days for each target date).
- **`X_seq_all`** ∈ ℝ^{N×w×n_exog}: sliding windows of the **scaled** exogenous features.
- **`y_tgt_all`** ∈ ℝ^{N}: the **next‑day** target (scaled) for each window.
- **`tgt_idx_all`**: timestamps (length `N`) for those targets (if a window spans `t−w..t−1`, the target index is `t`).
- **Calendar split**: `train_end_date`, `test_start_date`, `test_end_date` (paper uses 2014–2017 in‑sample, 2018 out‑of‑sample).
- **`y_scaler`**: `StandardScaler` fitted on the **in‑sample** target, to invert predictions to original units.
- **Refit cadence & training params**: `refit_every`, `epochs_per_refit`, `batch_size`, `val_fraction`.

---

#### What happens step‑by‑step
1) **Pick windows for initial training & test**
   - `initial_train_mask`: windows with **target date** ≤ `train_end_date`.
   - `test_mask`: windows with **target date** in `[test_start_date, test_end_date]`.
   - `pred_idx`: the ordered list of test target timestamps.

2) **Build the model**
   - `build_cnn_mlp(window, n_exog)` creates two branches (CNN on y, MLP on X) and concatenates them.

3) **Helper `fit_up_to(date_cutoff)`**
   - Trains on **all windows** with target date ≤ `date_cutoff`.
   - Uses a simple **time‑based validation**: last `val_fraction` is validation; the rest is training.

4) **Initial fit**
   - `fit_up_to(train_end_date)` trains once on the full in‑sample period.

5) **Walk forward through test dates**
   - For each test date `t` in `pred_idx`:
     - **Periodic refit**: if `i % refit_every == 0` and `t > last_refit_date`, call `fit_up_to(t−1)` to include all data up to **yesterday**.
     - **Find the right window**: locate the row where `target_index == t` (via `tgt_lut`) — this window ends at `t−1`.
     - **Predict (scaled)**: `yhat_scaled = model.predict([y_seq[row], X_seq[row]])`.
     - **Invert scaling**: `yhat = μ_y + σ_y · yhat_scaled` using the **training‑only** `y_scaler`.
     - Store `yhat` aligned to timestamp `t`.

6) **Return**
   - A `pd.Series(preds, index=pred_idx)` of forecasts in **original units**.

---

#### Tiny numeric example (window = 3, hop = 1)
Assume we focus on 4 test days: **2018‑01‑01 … 2018‑01‑04**. Values below are already **scaled** (zero‑mean, unit‑var).

##### `y_seq_all` (target windows, shape `N×3×1`)

- **2018‑01‑02**:
  - Not a refit point if `refit_every=7`.
  - Row for 2018‑01‑02 → output `yhat_scaled = 0.28` → `yhat = 100 + 20·0.28 = 105.6`.

This continues through the test period; at each **refit boundary** the training set expands to include all windows whose targets are ≤ **yesterday**, maintaining causality.

````python
# ---------------------------------------------------
# 8) Metrics (RMSE, MAE, sMAPE, MASE with m=7)
# ---------------------------------------------------
def rmse(y_true, y_pred): 
    return mean_squared_error(y_true, y_pred, squared=False)

def smape(y_true, y_pred, eps=1e-8):
    y_true = np.asarray(y_true, float); y_pred = np.asarray(y_pred, float)
    return np.mean(2*np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred) + eps)) * 100.0

def mase(y_true, y_pred, insample, m=7, eps=1e-8):
    insample = np.asarray(insample, float)
    if len(insample) <= m: 
        return np.nan
    denom = np.mean(np.abs(insample[m:] - insample[:-m]))
    return np.mean(np.abs(np.asarray(y_true) - np.asarray(y_pred))) / (denom + eps)

# Align truth
y_test = y_all.loc[cnn_mlp_pred.index]
# In-sample (levels) for MASE denominator
insample_levels = y_all.loc[(y_all.index >= ins_start) & (y_all.index <= ins_end)]

metrics = {
    "RMSE":  rmse(y_test.values, cnn_mlp_pred.values),
    "MAE":   mean_absolute_error(y_test.values, cnn_mlp_pred.values),
    "sMAPE%": smape(y_test.values, cnn_mlp_pred.values),
    "MASE":  mase(y_test.values, cnn_mlp_pred.values, insample_levels.values, m=7),
}
print("CNN+MLP (w=30, hop=1, expanding weekly) — Test 2018 metrics:")
for k,v in metrics.items():
    print(f"  {k}: {v:.3f}")

# ---------------------------------------------------
# 9) Plot
# ---------------------------------------------------
plt.figure(figsize=(14,5))
plt.plot(y_test.index, y_test.values, label="Actual")
plt.plot(cnn_mlp_pred.index, cnn_mlp_pred.values, label="CNN+MLP forecast")
plt.title("Total Store Sales — 1D-CNN (y) + MLP (exog) — Expanding Window (2018)")
plt.xlabel("Date")
plt.ylabel("Units")
plt.legend()
plt.grid(True)
plt.show()
````

## Forecasting Model Selection **F\*** in Figure 2 (Model Selection vs. NND)

**Short answer.** In Figure 2, **F\*** is the **base forecasting model chosen by model selection** (expanding‑window CV) from a pool of **standard methods** (e.g., Naive, ARIMA/ETS/NAR, **ARIMAX**, **NARX**, and simple/weighted combinations). ---

### What **F\*** stands for
- **F\*** = “the forecasting model used at that node.”
- Whenever the pipeline needs a **base forecast**, the authors **select F\*** by **expanding‑window CV** from the **standard pool** (ARIMAX, NARX, ARIMA/ETS/NAR, etc.).

### Where the CNN+MLP (NND) fits
- The **NND** is a **two‑branch deep model**:
  - **1D‑CNN** over a fixed window of the series (e.g., \(w=30\) days) to capture short‑term temporal patterns.
  - **MLP** over exogenous features (promotion signal and calendar dummies).
  - The two streams are **concatenated** and mapped to the forecast.
- In **standard top‑down**, a selected **F\*** produces the **top‑level forecast**; then **NND** disaggregates that forecast **down the hierarchy**, enforcing coherence.
- In **iterative top‑down**, the **NND itself** can play the role of **F\*** **between adjacent levels** (e.g., level \(k\!- 1\) → level \(k\)). At the very top level, **F\*** is still selected among the standard forecasting methods.

### Did model selection for **F\*** include ARIMAX/NARX/ETS?
**Yes.** The model‑selection step explicitly considers **ARIMAX/NARX/ETS/ARIMA/NAR** (and combinations) as candidates. The best performer (per expanding‑window CV on the training period) is chosen as **F\*** wherever a base forecast is required.

### Full workflow (how the pieces interact)
1. **Prepare features & windows**: promotions (binary or share) + DOW & Month dummies; build fixed windows (e.g., \(w=30\), hop=1).
2. **Model selection for F\***: on the **training years** (2014–2017), run **expanding‑window CV** over standard candidates (ARIMAX/NARX/ETS/etc.) and pick the best as **F\***.
3. **Train final components**:
   - **F\***: retrain the selected base model on all in‑sample data at the relevant level(s).
   - **NND (CNN+MLP)**: train to **disaggregate** forecasts (standard top‑down) or to predict **between adjacent levels** (iterative top‑down), using the same 30‑day windows and exogenous inputs.
4. **Forecast & reconcile**: apply **F\*** to produce base forecasts, use **NND** to propagate/disaggregate, and ensure **hierarchical coherence** (built into NND’s training objective).
5. **Evaluate**: report MASE/SMAPE/RMSE on the held‑out year (e.g., 2018) and compare the proposed NND against ARIMAX/NARX/ETS baselines and their combinations.

**Bottom line:** **F\*** is **selected** from **standard forecasting competitors** (ARIMAX/NARX/ETS, etc.), while the **1D‑CNN + MLP (NND)** is the paper’s **disaggregation/forecasting** component used to move information down the hierarchy (and optionally between levels in the iterative variant). They are **distinct** roles in the overall pipeline.

````python
# ---------------------------------------------------------------
# 1) Utilities to extract brand & item series and exogenous features
# ---------------------------------------------------------------

def get_items_for_brand(brand: str) -> List[str]:
    items = brand_dict.get(brand, [])
    if not items:
        raise ValueError(f"No items found for brand {brand} in brand_dict.")
    return items


def get_brand_series(frame: pd.DataFrame, brand: str) -> pd.Series:
    # If a precomputed brand column exists, prefer it
    col = f"Brand_QTY_{brand}"
    if col in frame.columns:
        y = frame[col].astype(float)
    else:
        # Sum item-level quantities
        items = get_items_for_brand(brand)
        qty_cols = [f"QTY_{brand}_{i}" for i in items if f"QTY_{brand}_{i}" in frame.columns]
        if not qty_cols:
            raise ValueError(f"No QTY columns for {brand} present in df.")
        y = frame[qty_cols].sum(axis=1).astype(float)
    return y.asfreq("D").fillna(0.0)


def build_brand_exog(frame: pd.DataFrame, brand: str) -> pd.DataFrame:
    ex = pd.DataFrame(index=frame.index)
    # Promo share = relative number of items under promo for this brand
    items = get_items_for_brand(brand)
    promo_cols = [f"PROMO_{brand}_{i}" for i in items if f"PROMO_{brand}_{i}" in frame.columns]
    if promo_cols:
        ex["promo_share"] = frame[promo_cols].sum(axis=1) / float(len(promo_cols))
    else:
        ex["promo_share"] = 0.0
    # Calendar dummies: DOW & Month (drop_first to avoid dummy trap if intercept exists)
    dow = pd.get_dummies(frame.index.dayofweek, prefix="dow", drop_first=True)
    mon = pd.get_dummies(frame.index.month,     prefix="mon", drop_first=True)
    dow.index = ex.index; mon.index = ex.index
    ex = pd.concat([ex, dow, mon], axis=1).astype(float)
    return ex.asfreq("D").fillna(0.0)


def build_item_exog(frame: pd.DataFrame, brand: str, items: List[str]) -> Dict[str, pd.DataFrame]:
    # Shared calendar dummies
    dow = pd.get_dummies(frame.index.dayofweek, prefix="dow", drop_first=True)
    mon = pd.get_dummies(frame.index.month,     prefix="mon", drop_first=True)
    dow.index = frame.index; mon.index = frame.index
    calendar = pd.concat([dow, mon], axis=1).astype(float)

    exog = {}
    for it in items:
        promo_col = f"PROMO_{brand}_{it}"
        if promo_col in frame.columns:
            promo = frame[[promo_col]].astype(float)
        else:
            promo = pd.DataFrame({promo_col: np.zeros(len(frame), dtype=float)}, index=frame.index)
        exog[it] = pd.concat([promo, calendar], axis=1).asfreq("D").fillna(0.0)
    return exog


def get_items_matrix(frame: pd.DataFrame, brand: str, items: List[str]) -> pd.DataFrame:
    cols = [f"QTY_{brand}_{it}" for it in items if f"QTY_{brand}_{it}" in frame.columns]
    if len(cols) != len(items):
        missing = [it for it in items if f"QTY_{brand}_{it}" not in frame.columns]
        raise ValueError(f"Missing QTY columns for items: {missing}")
    return frame[cols].astype(float).asfreq("D").fillna(0.0)
````

### Window Functions

````python
# ---------------------------------------------------------------
# 2) Window builders
# ---------------------------------------------------------------

def build_brand_sequences(y: pd.Series, X: pd.DataFrame, window: int = 30, hop: int = 1,
                          scaler_y: StandardScaler = None, scaler_X: StandardScaler = None):
    """Return y_seq (N,w,1), X_seq (N,w,nx), y_tgt (N,), idx (N,) — scaled via provided scalers.
    Fits scalers if None, using only the rows whose target index is <= TRAIN_END.
    """
    assert y.index.equals(X.index)
    # Build raw sequences first (unscaled)
    vals_y = y.values.astype(float)
    vals_X = X.values.astype(float)
    n, nx = len(y), X.shape[1]

    y_seq_list, X_seq_list, y_tgt_list, idx_list = [], [], [], []
    t = window
    while t < n:
        y_seq_list.append(vals_y[t-window:t].reshape(window, 1))
        X_seq_list.append(vals_X[t-window:t].reshape(window, nx))
        y_tgt_list.append(vals_y[t])
        idx_list.append(y.index[t])
        t += hop

    y_seq = np.stack(y_seq_list, axis=0)
    X_seq = np.stack(X_seq_list, axis=0)
    y_tgt = np.array(y_tgt_list, dtype=float)
    idx = pd.to_datetime(pd.Index(idx_list))

    # Fit scalers on in-sample targets (up to TRAIN_END)
    if scaler_y is None:
        scaler_y = StandardScaler()
        ins_mask = (idx <= TRAIN_END)
        scaler_y.fit(y_tgt[ins_mask].reshape(-1,1))
    if scaler_X is None:
        scaler_X = StandardScaler()
        # To fit X scaler, take all windows whose target <= TRAIN_END
        ins_mask = (idx <= TRAIN_END)
        X_ins = X_seq[ins_mask].reshape(ins_mask.sum()*window, nx)
        scaler_X.fit(X_ins)

    # Apply scaling
    y_tgt_sc = scaler_y.transform(y_tgt.reshape(-1,1)).ravel()
    X_seq_sc = scaler_X.transform(X_seq.reshape(-1, nx)).reshape(-1, window, nx)
    y_seq_sc = scaler_y.transform(y_seq.reshape(-1,1)).reshape(-1, window, 1)

    return y_seq_sc, X_seq_sc, y_tgt_sc, idx, scaler_y, scaler_X


def build_item_disagg_sequences(
    y_brand: pd.Series,
    Y_items: pd.DataFrame,
    item_exog: Dict[str, pd.DataFrame],
    window: int = 30,
    hop: int = 1,
) -> Tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray, pd.DatetimeIndex]:
    """
    Build sequences for disaggregation model.
    Returns:
      - yB_seq:   (N, w, 1)    brand history (UNSCALED)
      - Xitem_seq:(N, I, w, m) per-item exog windows (UNSCALED)
      - yB_t:     (N,)         brand scalar target at t (UNSCALED)
      - Yitem_t:  (N, I)       item targets at t (UNSCALED)
      - idx:      (N,)         target timestamps
    """
    items = [c.split("_")[-1] for c in Y_items.columns]
    n_items = len(items)

    # infer item exog width (all items share schema)
    any_item = items[0]
    m = item_exog[any_item].shape[1]

    vals_yB = y_brand.values.astype(float)
    vals_items = Y_items.values.astype(float)  # shape (T, I)
    n = len(y_brand)

    yB_seq_list, Xitem_seq_list, yB_t_list, Yitem_t_list, idx_list = [], [], [], [], []

    t = window
    while t < n:
        # brand history
        yB_seq_list.append(vals_yB[t-window:t].reshape(window, 1))
        # per-item exog windows
        per_item_windows = []
        for it in items:
            Xi = item_exog[it].values.astype(float)
            per_item_windows.append(Xi[t-window:t].reshape(window, m))
        Xitem_seq_list.append(np.stack(per_item_windows, axis=0))  # (I, w, m)
        # targets
        yB_t_list.append(vals_yB[t])
        Yitem_t_list.append(vals_items[t, :])
        idx_list.append(y_brand.index[t])
        t += hop

    yB_seq   = np.stack(yB_seq_list, axis=0)                      # (N, w, 1)
    Xitem_seq= np.stack(Xitem_seq_list, axis=0)                   # (N, I, w, m)
    yB_t     = np.array(yB_t_list, dtype=float)                   # (N,)
    Yitem_t  = np.stack(Yitem_t_list, axis=0).astype(float)       # (N, I)
    idx      = pd.to_datetime(pd.Index(idx_list))

    return yB_seq, Xitem_seq, yB_t, Yitem_t, idx
````

## ML Model

### 1D CNN + MLP & NDD Defn

````python
# ---------------------------------------------------------------
# 3) Models
# ---------------------------------------------------------------

def build_brand_cnn_mlp(window: int, n_exog: int) -> Model:
    # y branch
    inp_y = layers.Input(shape=(window, 1), name="y_seq")
    a = layers.Conv1D(16, 3, padding="causal", activation="relu")(inp_y)
    a = layers.Conv1D(16, 3, padding="causal", activation="relu")(a)
    a = layers.GlobalAveragePooling1D()(a)
    # X branch
    inp_x = layers.Input(shape=(window, n_exog), name="x_seq")
    b = layers.Flatten()(inp_x)
    b = layers.Dense(64, activation="relu")(b)
    # fuse
    h = layers.Concatenate()([a, b])
    h = layers.Dense(64, activation="relu")(h)
    out = layers.Dense(1, name="yhat_scaled")(h)  # trained in scaled space
    model = Model([inp_y, inp_x], out)
    model.compile(optimizer="adam", loss="mse")
    return model
````

````python
#    NDD 

def build_disagg_nnd(window: int, n_items: int, n_item_exog: int,
                     brand_latent_dim: int = 16) -> Model:
    """NND that outputs per-item forecasts coherent with the brand scalar.
    Inputs:
      - yB_seq: (w,1) brand history
      - Xitem_seq: (I,w,m) per-item exog sequence
      - yB_scalar: (1,) brand scalar at target time (true for training, forecast at inference)
    Output:
      - Yitem_pred: (I,) per-item forecasts that sum to yB_scalar (softmax shares * scalar)
    """
    # Brand branch
    inp_yB = layers.Input(shape=(window, 1), name="yB_seq")
    a = layers.Conv1D(brand_latent_dim, 3, padding="causal", activation="relu")(inp_yB)
    a = layers.GlobalAveragePooling1D()(a)  # (batch, brand_latent_dim)

    # Per-item exog branch
    inp_Xi = layers.Input(shape=(n_items, window, n_item_exog), name="Xitem_seq")
    # Flatten time+features for each item, keep item axis
    z = layers.Reshape((n_items, window * n_item_exog))(inp_Xi)
    z = layers.TimeDistributed(layers.Dense(64, activation="relu"))(z)
    z = layers.TimeDistributed(layers.Dense(32, activation="relu"))(z)  # (batch, I, 32)

    # Repeat brand latent across items and fuse
    a_rep = layers.RepeatVector(n_items)(a)  # (batch, I, brand_latent_dim)
    u = layers.Concatenate(axis=-1)([z, a_rep])  # (batch, I, 32+latent)
    u = layers.TimeDistributed(layers.Dense(32, activation="relu"))(u)
    logits = layers.TimeDistributed(layers.Dense(1))(u)  # (batch, I, 1)
    logits = layers.Reshape((n_items,))(logits)           # (batch, I)
    shares = layers.Softmax(axis=-1, name="shares")(logits)  # sums to 1 across items

    # Multiply by brand scalar to get coherent item forecasts
    inp_yB_scalar = layers.Input(shape=(1,), name="yB_scalar")  # (batch,1)
    yB_rep = layers.Concatenate()([inp_yB_scalar for _ in range(n_items)])  # (batch, I)
    y_items_pred = layers.Multiply(name="Yitem_pred")([shares, yB_rep])

    model = Model([inp_yB, inp_Xi, inp_yB_scalar], y_items_pred)
    model.compile(optimizer="adam", loss="mse")  # MSE on items in original units
    return model
````

### Explained: Coherence in the Brand→Items Forecast (Why item predictions sum to the brand)

##### What “coherence” means here
Coherence means that, at each forecast date \(t\), the **sum of all item-level predictions equals the brand-level prediction**:
$$
\sum_{i=1}^{I} \hat{y}_{i}(t) \;=\; \hat{y}_{\text{brand}}(t).
$$

---

##### Where coherence is enforced

###### 1) Inside the disaggregation network (NND)
The NND produces **item shares** via a Softmax and then **scales** them by the **brand scalar** (the brand’s forecast at time $(t)$):

- **Shares:** the network outputs logits per item, then applies a Softmax across items.  
  Result: $(\text{shares}_i(t) \ge 0)$ and $(\sum_i \text{shares}_i(t) = 1)$.

  ```python
logits = layers.TimeDistributed(layers.Dense(1))(u)   # (batch, I, 1)
logits = layers.Reshape((n_items,))(logits)           # (batch, I)
shares = layers.Softmax(axis=-1, name="shares")(logits)  # sums to 1 across items
  ```

- **Scaling:** item forecasts are computed as  
  $$
  \hat{y}_{i}(t) = \text{shares}_{i}(t) \times \hat{y}_{\text{brand}}(t).
  $$

```python
inp_yB_scalar = layers.Input(shape=(1,), name="yB_scalar")   # (batch, 1)
yB_rep = layers.Concatenate()([inp_yB_scalar for _ in range(n_items)])  # (batch, I)
y_items_pred = layers.Multiply(name="Yitem_pred")([shares, yB_rep])     # (batch, I)
```

Therefore,
$$
\sum_i \hat{y}_{i}(t) \;=\; \hat{y}_{\text{brand}}(t)\,\sum_i \text{shares}_{i}(t)
\;=\; \hat{y}_{\text{brand}}(t)\times 1
\;=\; \hat{y}_{\text{brand}}(t).
$$
This guarantees **coherence by construction** (no post-hoc fix needed).

###### 2) At inference time: feeding the brand forecast into the NND
During testing, the **brand forecast series** \(\hat{y}_{\text{brand}}\) for each timestamp is passed to the NND as the **brand scalar** input. Because item predictions are $(\text{shares} \times \hat{y}_{\text{brand}})$, their sum automatically equals that brand forecast at every timestamp.

```python
# Build brand scalars for test from brand forecast (ensures coherence)
yB_scalar_test = brand_pred_series.reindex(idx_d[d_test]).values.reshape(-1,1)
X_disagg_test[2] = yB_scalar_test

# Predict per-item test forecasts (already coherent with brand)
Y_items_pred = disagg_model.predict(X_disagg_test, verbose=0)
```

###### 3) The coherence printout is only a **check**
After prediction, the code computes
```python
sum_items_pred = items_pred_df.sum(axis=1)
coh_max_abs_err = np.max(np.abs(sum_items_pred.values
                   - brand_pred_series.reindex(sum_items_pred.index).values))
print(f"Max abs coherence error (should be ~0): {coh_max_abs_err:.6f}")
```

$$
\max_t \left| \sum_i \hat{y}_i(t) - \hat{y}_{\text{brand}}(t) \right|
$$
and prints it. This **verifies** the property (should be ~0 aside from tiny floating-point error); it does **not** enforce coherence.

---

##### Tiny numeric example (2 items)

Suppose at one test date \(t\):

- Brand forecast: \(\hat{y}_{\text{brand}}(t) = 100\).
- NND logits for two items: \([1.5,\; 0.0]\).

Softmax shares:
$$
\text{shares}_1 = \frac{e^{1.5}}{e^{1.5}+e^{0}} \approx 0.8176, \quad
\text{shares}_2 = 1 - 0.8176 = 0.1824.
$$

Item forecasts:
$$
\hat{y}_1 = 0.8176 \times 100 = 81.76, \quad
\hat{y}_2 = 0.1824 \times 100 = 18.24.
$$

Sum:
$$
\hat{y}_1 + \hat{y}_2 = 81.76 + 18.24 = 100.00 = \hat{y}_{\text{brand}}.
$$

That’s coherence.

---

##### Training vs. inference (why it still holds)
- **Training:** the NND uses the **true** brand scalar $(y_{\text{brand}}(t))$ so it learns to map history + exogenous inputs to **shares**.
- **Inference:** the NND uses the **predicted** brand scalar $(\hat{y}_{\text{brand}}(t))$. Because items are computed as $(\text{shares}\times \hat{y}_{\text{brand}})$, the **sum of item forecasts equals the brand forecast** at each test timestamp.

> **Tip:** Ensure your brand forecast covers every test timestamp you disaggregate. If some dates are missing, reindexing will introduce NaNs and the coherence check will spike.

````python
#-----------------------------------------------
#          Plotting Brand  (Train + Test, Forecast)
#-----------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from typing import Union

def plot_brand_timeseries_and_forecast(
    df: pd.DataFrame,
    brand: str,
    brand_pred,                       # pd.Series OR array-like
    *,
    pred_index: Union[pd.Index, list, None] = None,  # required if brand_pred is array-like
    date_col_indexed: bool = True,
    start: Union[str, pd.Timestamp, None] = None,
    end: Union[str, pd.Timestamp, None] = None,
    train_end: Union[str, pd.Timestamp, None] = None,
    forecast_label: str = "NDD/F* forecast",
    title_suffix: str = ""
) -> pd.Series:
    """
    Plot brand-level actual time series and the provided forecast.

    Parameters
    ----------
    df : DataFrame
        Your full data. Either already indexed by DATE (DatetimeIndex), or with a 'DATE' column.
    brand : str
        Brand code, e.g. "B1".
    brand_pred : pd.Series
        Forecasted brand series in ORIGINAL UNITS (not scaled), indexed by dates to plot.
        e.g., the brand scalar produced by your base forecaster (F*) used by NDD.
    date_col_indexed : bool, default True
        If False, the function will set 'DATE' as index (and to_datetime) before proceeding.
    start, end : str | Timestamp | None
        Optional date limits for the background actual time series.
    train_end : str | Timestamp | None
        Optional vertical line to indicate end of training period.
    forecast_label : str
        Legend label for the forecast curve.
    title_suffix : str
        Extra text appended to the title.

    Returns
    -------
    pd.Series
        The actual brand series aligned to the forecast index (useful for scoring).
    """
    # Ensure index is a DatetimeIndex
    if not date_col_indexed:
        if "DATE" not in df.columns:
            raise ValueError("DATE column not found; set date_col_indexed=True or add DATE.")
        df = df.copy()
        df["DATE"] = pd.to_datetime(df["DATE"])
        df = df.set_index("DATE").sort_index()

    if not isinstance(df.index, pd.DatetimeIndex):
        raise ValueError("df must be indexed by DatetimeIndex or pass date_col_indexed=False.")

    # Build actual brand series
    brand_sum_col = f"Brand_QTY_{brand}"
    if brand_sum_col in df.columns:
        y_brand = pd.to_numeric(df[brand_sum_col], errors="coerce").fillna(0.0)
    else:
        qty_cols = [c for c in df.columns if c.startswith(f"QTY_{brand}_")]
        if not qty_cols:
            raise ValueError(f"No columns found for brand {brand} (looked for {brand_sum_col} or QTY_{brand}_*).")
        y_brand = df[qty_cols].apply(pd.to_numeric, errors="coerce").fillna(0.0).sum(axis=1)

    # Optional view window for background series
    y_plot = y_brand.loc[start:end] if (start or end) else y_brand

    # --- Coerce brand_pred to a Series with a Date index ---
    if isinstance(brand_pred, pd.Series):
        pred_series = brand_pred.copy()
        if not isinstance(pred_series.index, pd.DatetimeIndex):
            raise ValueError("brand_pred Series must have a DatetimeIndex.")
        pred_series = pred_series.sort_index()
    else:
        # array-like -> need pred_index
        if pred_index is None:
            raise ValueError("brand_pred is array-like; supply pred_index (list/Index of timestamps).")
        arr = np.asarray(brand_pred).reshape(-1)
        idx = pd.to_datetime(pd.Index(pred_index))
        if len(arr) != len(idx):
            raise ValueError(f"Length mismatch: brand_pred={len(arr)} vs pred_index={len(idx)}.")
        pred_series = pd.Series(arr, index=idx, name="forecast").sort_index()

    # Align actuals to forecast index for metrics
    y_true_on_pred = y_brand.reindex(pred_series.index)

    # Plot
    plt.figure(figsize=(14, 5))
    plt.plot(y_plot.index, y_plot.values, label=f"Actual {brand}", alpha=0.9)
    plt.plot(pred_series.index, pred_series.values, label=f"{forecast_label} {brand}", linewidth=2)
    if train_end is not None:
        plt.axvline(pd.to_datetime(train_end), linestyle="--", label="Train end")
    plt.title(f"{brand} — Actual vs Forecast {title_suffix}".strip())
    plt.xlabel("Date")
    plt.ylabel("Units")
    plt.grid(True)
    plt.legend()
    plt.show()

    return y_true_on_pred
````

````python
# ---------------------------------------------------------------
# 6.2) Plotting helper — Actual vs Forecast for a specific ITEM
# ---------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from typing import Union

def plot_item_test_overlay(
    df: pd.DataFrame,
    brand: str,
    item: Union[str, int],
    test_start: Union[str, pd.Timestamp],
    test_end: Union[str, pd.Timestamp],
    item_pred: Union[pd.Series, pd.DataFrame, np.ndarray],  # pd.Series OR pd.DataFrame (with the item col) OR array-like
    pred_index: Union[pd.Index, list, None] = None,      # required only if item_pred is array-like
    date_col_indexed: bool = True,
    y_label: str = "Units",
    actual_label: str = "Actual",
    forecast_label: str = "Forecast",
    title_suffix: str = "(Test Period)"
) -> tuple[pd.Series, pd.Series]:
    """
    Overlay Actual vs Forecast for a chosen item (brand+item) on ONE chart, restricted to [test_start, test_end].

    item_pred can be:
      • pd.Series with a DatetimeIndex
      • pd.DataFrame that contains the column f"QTY_{brand}_{item}"
      • array-like (then pass pred_index with matching timestamps)

    Returns (y_true_test, y_pred_on_true_index).
    """

    # Actual item series
    item_str = str(item)
    item_col = f"QTY_{brand}_{item_str}"
    if item_col not in df.columns:
        raise ValueError(f"Actuals column not found: {item_col}")

    y_item = pd.to_numeric(df[item_col], errors="coerce").fillna(0.0)

    # Restrict to test window
    test_start = pd.to_datetime(test_start)
    test_end   = pd.to_datetime(test_end)
    y_true_test = y_item.loc[test_start:test_end]
    if y_true_test.empty:
        raise ValueError("Actual item test window is empty; verify test_start/test_end against your data range.")

    # Forecast series coercion
    if isinstance(item_pred, pd.Series):
        if not isinstance(item_pred.index, pd.DatetimeIndex):
            raise ValueError("item_pred Series must have a DatetimeIndex.")
        y_pred_test = item_pred.sort_index().loc[test_start:test_end]

    elif isinstance(item_pred, pd.DataFrame):
        # Expect the item column inside the DataFrame (e.g., items_pred_df)
        if item_col not in item_pred.columns:
            raise ValueError(f"item_pred DataFrame does not contain column {item_col}.")
        pred_series = item_pred[item_col]
        if not isinstance(pred_series.index, pd.DatetimeIndex):
            raise ValueError("item_pred DataFrame must have a DatetimeIndex.")
        y_pred_test = pred_series.sort_index().loc[test_start:test_end]

    else:
        # array-like → need pred_index
        if pred_index is None:
            raise ValueError("item_pred is array-like; supply pred_index (timestamps).")
        arr = np.asarray(item_pred).reshape(-1)
        idx = pd.to_datetime(pd.Index(pred_index))
        if len(arr) != len(idx):
            raise ValueError(f"Length mismatch: item_pred={len(arr)} vs pred_index={len(idx)}.")
        pred_series = pd.Series(arr, index=idx, name=item_col).sort_index()
        y_pred_test = pred_series.loc[test_start:test_end]

    if y_pred_test.empty:
        raise ValueError("Forecast item test window is empty; check forecast dates vs test window.")

    # Align forecast to actual dates (handy for metrics)
    y_pred_on_true_index = y_pred_test.reindex(y_true_test.index)

    # Plot (single figure with both lines)
    plt.figure(figsize=(14, 4))
    plt.plot(y_true_test.index, y_true_test.values, label=f"{actual_label} {brand}-{item_str}", linewidth=2)
    plt.plot(y_pred_test.index, y_pred_test.values, label=f"{forecast_label} {brand}-{item_str}",
             linewidth=2, linestyle="--")
    plt.title(f"{brand}-{item_str} — {actual_label} vs {forecast_label} {title_suffix}".strip())
    plt.xlabel("Date")
    plt.ylabel(y_label)
    plt.grid(True)
    plt.legend()
    plt.show()

    return y_true_test, y_pred_on_true_index
````
